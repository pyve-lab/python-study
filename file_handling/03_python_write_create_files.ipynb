# 파이썬 파일 쓰기 (Python File Writing)

## 기존 파일에 쓰기

기존 파일에 내용을 쓰려면 `open()` 함수에 매개변수를 추가해야 합니다.

* **`"a"` (추가)**: 파일 끝에 내용을 추가합니다.
* **`"w"` (쓰기)**: 기존 내용을 덮어씁니다.

예시: 파일 끝에 내용 추가 (`"a"`)

```python
# "demofile.txt" 파일을 열고 파일에 내용을 추가하세요.
with open("demofile.txt", "a") as f:
  f.write("Now the file has more content!")

#"파일 추가(appending) 후 파일을 열고 읽습니다:
with open("demofile.txt") as f:
  print(f.read())
```

## 기존 콘텐츠 덮어쓰기
파일의 기존 내용을 덮어쓰려면 "w" 매개변수를 사용하십시오.

예시: 기존 내용 덮어쓰기 ("w")
"demofile.txt" 파일을 열고 내용을 덮어씁니다.

```python
# "demofile.txt" 파일을 열고 내용을 덮어쓰세요.
with open("demofile.txt", "w") as f:
  f.write("Woops! I have deleted the content!")

#덮어쓰기 후 파일을 열고 읽습니다:
with open("demofile.txt") as f:
  print(f.read())
```

참고: "w" 메서드는 파일 전체를 덮어씁니다.

## 새 파일 만들기

파이썬에서 새 파일을 생성하려면 open()다음 매개변수 중 하나를 사용하여 해당 메서드를 사용합니다.

* **"x"- 생성**: 파일을 생성하고, 파일이 이미 존재하는 경우 오류를 반환합니다.

* **"a"- 추가**: 지정된 파일이 존재하지 않으면 파일을 생성합니다.

* **"w"- 쓰기**: 지정된 파일이 존재하지 않으면 파일을 생성합니다.

예시: "myfile.txt"라는 이름의 새 파일을 만드세요:

```python
f = open("myfile.txt", "x")
```

결과: 새로운 빈 파일이 생성됩니다.

참고: 파일이 이미 존재하는 경우 오류가 발생합니다.
