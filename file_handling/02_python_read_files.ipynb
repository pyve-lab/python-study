# 파이썬 파일 열기 (Python File Open)

## 서버에서 파일 열기

파이썬 파일이 위치한 폴더와 **같은 폴더**에 다음과 같은 파일이 있다고 가정해 보겠습니다.

**demofile.txt 내용:**
```txt
Hello! Welcome to demofile.txt
This file is for testing purposes.
Good Luck!


파일을 열려면 파이썬의 내장함수 `open()` 를 사용하십시오.

이 `open()` 함수는 파일 객체를 반환하며, 이 객체에는 파일 내용을 읽는 `read()` 메서드가 있습니다.


````python
# 예시
f = open("demofile.txt")
print(f.read())
````

````python
# 예시 
# 다른 위치에 있는 파일을 엽니다.
# f = open("D:\\myfiles\welcome.txt")
f = open("C:\\myfiles\welcome.txt") # 실제 pc에 이경로로 지정해봄
print(f.read())
````

## with 문 사용

파일을 열 때 with 문을 사용할 수도 있습니다.

````python
# 예시
with open("demofile.txt") as f:
  print(f.read())
````

그러면 파일을 닫는 것에 대해 걱정할 필요가 없습니다. 해당 with명령문이 알아서 처리해 줍니다.

## 파일 닫기

파일 작업을 마친 후에는 항상 파일을 닫는 것이 좋습니다.

해당 구문을 사용하지 않는 경우 with, 파일을 닫기 위해 close 문을 작성해야 합니다.

````python
# 예시
f = open("demofile.txt")
print(f.readline())
f.close()
````

**참고** : 파일을 열 때는 항상 닫아야 합니다. 버퍼링 문제로 인해 파일을 닫기 전까지는 변경 사항이 반영되지 않을 수 있습니다.

## 파일의 일부만 읽기 (Read Only Parts of the File)

기본적으로 read() 메서드는 전체 텍스트를 반환하지만, 반환할 문자 수를 지정할 수도 있습니다.

````python
# 예시
# 파일의 처음 5개 문자를 반환합니다.

with open("demofile.txt") as f:
  print(f.read(5))
````

## 줄 단위로 읽기 (Read Lines)

readline() 메서드를 사용하여 한 줄을 반환할 수 있습니다.

````python
# 예시
# 파일에서 한 줄을 읽으세요

with open("demofile.txt") as f:
  print(f.readline())
````

readline()을 두 번 호출하면 처음 두 줄을 읽을 수 있습니다.

````python
# 예시
# 파일에서 두 줄을 읽으세요
with open("demofile.txt") as f:
  print(f.readline())
  print(f.readline())
````

파일의 각 줄을 순회(looping)하면 파일 전체를 한 줄씩 읽을 수 있습니다.

````python
# 예시
# 파일을 한 줄씩 순회합니다
with open("demofile.txt") as f:
  for x in f:
    print(x)
````
