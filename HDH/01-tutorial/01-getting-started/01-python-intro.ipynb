### Python이란?

1991년 Guido van Rossum에 의해 만들어진 프로그래밍 언어. 웹 개발, 소프트웨어 개발, 수학 연산, 시스템 스크립팅 등에 사용됨.

### Python의 특징

- 크로스 플랫폼 지원 (Windows, Mac, Linux 등)
- 간결하고 읽기 쉬운 문법
- 인터프리터 방식으로 빠른 프로토타이핑 가능
- 들여쓰기로 스코프 정의 (중괄호 대신)
- 절차적, 객체지향, 함수형 프로그래밍 모두 지원

## Java와의 비교

### 문법 차이

**Python:**
```python
print("Hello, World!")

name = "Python"
age = 30

def greet(name):
    return f"Hello, {name}!"

print(greet("Developer"))
```

**Java:**
```java
public class Main {
    public static void main(String[] args) {
        System.out.println("Hello, World!");
        
        String name = "Java";
        int age = 30;
        
        System.out.println(greet("Developer"));
    }
    
    public static String greet(String name) {
        return "Hello, " + name + "!";
    }
}
```

### 개념적 차이

- **타입**: Python은 동적 타입, Java는 정적 타입
- **실행 방식**: Python은 인터프리터, Java는 컴파일 후 실행
- **스코프**: Python은 들여쓰기, Java는 중괄호
- **코드량**: Python이 더 간결
- **패러다임**: Python은 다중 패러다임, Java는 객체지향 중심
