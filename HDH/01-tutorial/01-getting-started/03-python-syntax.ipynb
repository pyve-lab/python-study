## 학습 내용

### Python Syntax 실행

- Command Line에서 직접 실행
- `.py` 파일로 작성 후 실행

### Python Indentation (들여쓰기)

Python에서 들여쓰기는 **매우 중요**. 다른 언어와 달리 들여쓰기가 코드 블록을 정의함.
- 최소 1칸 이상의 공백 필요 (일반적으로 4칸 사용)
- 같은 블록 내에서는 동일한 들여쓰기 사용 필수

### Python Variables

변수는 값을 할당할 때 자동으로 생성됨. 별도의 선언 명령어 없음.

### Comments (주석)

`#` 기호로 주석 처리. `#` 이후의 내용은 주석으로 처리됨.

### Statements (문장)

프로그램의 지시사항. Python에서는 세미콜론(`;`)이 선택사항이며 거의 사용하지 않음.


## Python 코드 실습


````python
# 들여쓰기로 코드 블록 정의
if 5 > 2:
    print("Five is greater than two!")
````

````python
# 변수는 값 할당 시 자동 생성
x = 5
y = "Hello, World!"
print(x, y)
````

````python
# 주석 예제
# 이것은 주석입니다
print("Hello, World!")  # 인라인 주석도 가능
````

````python
# 여러 문장 실행 (세미콜론 불필요)
print("Hello World!")
print("Have a good day.")
print("Learning Python is fun!")
````

## Java와의 비교

### 문법 구조 차이

**Python:**
```python
# 단순 스크립트로 시작
if 5 > 2:
    print("Five is greater than two!")

x = 5
y = "Hello, World!"

# 주석
print("Hello, World!")
```

**Java:**
```java
// 클래스 구조 필수
public class Main {
    public static void main(String[] args) {
        if (5 > 2) {
            System.out.println("Five is greater than two!");
        }
        
        int x = 5;
        String y = "Hello, World!";
        
        // 주석
        System.out.println("Hello, World!");
    }
}
```

### 개념적 차이

- **구조**: Python은 단순 스크립트, Java는 클래스 구조 필수
- **들여쓰기**: Python은 들여쓰기로 스코프 정의 (필수), Java는 중괄호 사용
- **세미콜론**: Python은 선택사항, Java는 필수
- **변수 선언**: Python은 값 할당 시 자동 생성, Java는 타입과 함께 명시적 선언
- **주석**: Python은 `#`, Java는 `//` 또는 `/* */`


## 정리

### 핵심 내용

1. Python의 들여쓰기는 코드 블록을 정의하는 필수 요소
2. 변수는 값 할당 시 자동으로 생성됨
3. 세미콜론은 선택사항이며 거의 사용하지 않음
4. Java와 달리 클래스 구조 없이 단순 스크립트로 시작 가능

### 느낀 점

- 들여쓰기가 필수, 코드 가독성 측면 장점.
- Java처럼 클래스 구조를 만들 필요 없이 바로 코드를 작성할 수 있어서 간편.
