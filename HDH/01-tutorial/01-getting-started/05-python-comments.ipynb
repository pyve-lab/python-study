## 학습 내용

### Comments (주석)의 용도

- 코드 설명
- 코드 가독성 향상
- 테스트 시 코드 실행 방지

### Creating a Comment

`#` 기호로 시작. `#` 이후의 내용은 Python이 무시함.

### Single-line Comments

한 줄 주석은 `#`로 시작. 줄 끝에 위치해도 가능.

### Multiline Comments

Python은 여러 줄 주석을 위한 특별한 문법이 없음. 두 가지 방법:
1. 각 줄 앞에 `#` 사용
2. Triple quotes (`"""` 또는 `'''`)로 감싼 문자열 리터럴 사용 (변수에 할당하지 않으면 Python이 무시하여 주석처럼 활용 가능)


## Python 코드 실습


````python
# 이것은 주석입니다
print("Hello, World!")
````

````python
# 줄 끝에 주석 추가
print("Hello, World!")  # 이것은 주석입니다
````

````python
# 코드 실행 방지 목적으로 주석 사용
# print("Hello, World!")
print("Cheers, Mate!")
````

````python
# 여러 줄 주석 방법 1: 각 줄에 # 사용
# 이것은 주석입니다
# 여러 줄에 걸쳐
# 작성할 수 있습니다
print("Hello, World!")
````

````python
# 여러 줄 주석처럼 사용: Triple quotes 문자열 리터럴 활용
# (변수에 할당하지 않으면 Python이 무시함)
"""
이것은 문자열 리터럴이지만
변수에 할당하지 않아서
주석처럼 활용할 수 있습니다
"""
print("Hello, World!")
````

````python
# Triple quotes를 변수에 할당하면 실제 문자열이 됨
message = """
이것은 문자열 리터럴이고
변수에 할당되어서
실제 문자열 값이 됩니다
"""
print(message)
````

## Java와의 비교

### 주석 방법 차이

**Python:**
```python
# 한 줄 주석
print("Hello, World!")

# 여러 줄 주석: 각 줄에 # 사용
# 이것은 주석입니다
# 여러 줄에 걸쳐 작성

# 문자열 리터럴을 주석처럼 활용
"""
이것은 문자열 리터럴이지만
변수에 할당하지 않아서
주석처럼 활용할 수 있습니다
"""
print("Hello, World!")
```

**Java:**
```java
// 한 줄 주석
System.out.println("Hello World");

// 여러 줄 주석
/* 이것은 주석입니다
   여러 줄에 걸쳐 작성 */
System.out.println("Hello World");
```

### 개념적 차이

- **한 줄 주석**: Python은 `#`, Java는 `//`
- **여러 줄 주석**: Python은 `#` 여러 번 사용, Java는 `/* */` 전용 문법
- **문자열 리터럴 활용**: Python은 triple quotes 문자열 리터럴을 주석처럼 활용 가능 (변수 할당 없이)
- **용도**: 둘 다 코드 설명, 가독성 향상, 코드 실행 방지 목적으로 사용
- **문법**: Python은 여러 줄 주석을 위한 특별한 문법이 없음


## 정리

### 핵심 내용

1. `#` 기호로 한 줄 주석 작성
2. 여러 줄 주석은 각 줄 앞에 `#` 사용
3. Triple quotes는 문자열 리터럴이지만, 변수에 할당하지 않으면 주석처럼 활용 가능
4. Python은 여러 줄 주석을 위한 특별한 문법이 없음

### 느낀 점

- Java의 `/* */`처럼 여러 줄 주석 전용 문법이 없지만, triple quotes 문자열 리터럴을 활용하면 비슷하게 작성할 수 있음.
- `#` 기호가 Java의 `//`보다 더 간단하고 직관적.
