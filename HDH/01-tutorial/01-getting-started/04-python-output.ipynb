## 학습 내용

### Print Text

`print()` 함수로 텍스트나 값을 출력할 수 있음. 여러 번 호출하면 기본적으로 각각 새 줄에 출력됨.

### Quotes (따옴표)

텍스트는 따옴표로 감싸야 함. `"` (double quotes) 또는 `'` (single quotes) 둘 다 사용 가능.

### Print Without a New Line

기본적으로 `print()`는 줄바꿈으로 끝남. 같은 줄에 출력하려면 `end` 파라미터 사용.

### Print Numbers

숫자는 따옴표 없이 출력 가능. `print()` 함수 내에서 수학 연산도 가능.

### Mix Text and Numbers

쉼표(`,`)로 텍스트와 숫자를 함께 출력 가능.


## Python 코드 실습


````python
# 텍스트 출력
print("Hello World!")
print("I am learning Python.")
print("It is awesome!")
````

````python
# Single quotes와 Double quotes 둘 다 사용 가능
print("This will work!")
print('This will also work!')
````

````python
# 같은 줄에 출력 (end 파라미터 사용)
print("Hello World!", end=" ")
print("I will print on the same line.")
````

````python
# 숫자 출력 (따옴표 없이)
print(3)
print(358)
print(50000)
````

````python
# 수학 연산
print(3 + 3)
print(2 * 5)
````

````python
# 텍스트와 숫자 함께 출력
print("I am", 35, "years old.")
````

## Java와의 비교

### 출력 방법 차이

**Python:**
```python
# 텍스트 출력
print("Hello World!")
print('This will also work!')

# 같은 줄에 출력
print("Hello World!", end=" ")
print("I will print on the same line.")

# 숫자 출력
print(3)
print(3 + 3)

# 텍스트와 숫자 함께
print("I am", 35, "years old.")
```

**Java:**
```java
// 텍스트 출력
System.out.println("Hello World!");
// System.out.println('This will not work!'); // 에러

// 같은 줄에 출력
System.out.print("Hello World! ");
System.out.print("I will print on the same line.");

// 숫자 출력
System.out.println(3);
System.out.println(3 + 3);

// 텍스트와 숫자 함께
System.out.println("I am " + 35 + " years old.");
```

### 개념적 차이

- **함수/메서드**: Python은 `print()`, Java는 `System.out.println()` 또는 `System.out.print()`
- **따옴표**: Python은 single/double quotes 둘 다 가능, Java는 double quotes만 가능
- **줄바꿈**: Python은 `end` 파라미터로 제어, Java는 `println()`(줄바꿈)과 `print()`(줄바꿈 없음) 구분
- **문자열 연결**: Python은 쉼표로 자동 공백 추가, Java는 `+` 연산자로 연결
