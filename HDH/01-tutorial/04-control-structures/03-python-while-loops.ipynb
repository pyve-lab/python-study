### Python Loops

Python에는 두 가지 기본적인 루프 명령이 있다:

- **while loops**: 조건이 참인 동안 문 집합을 실행
- **for loops**: (다음 챕터에서 학습)

### The while Loop

- `while` 루프를 사용하면 조건이 참인 동안 문 집합을 실행할 수 있음
- 루프는 조건이 False가 될 때까지 계속 실행됨
- **중요**: 조건에서 사용하는 변수를 증가시키는 것을 잊지 말아야 함. 그렇지 않으면 루프가 영원히 계속됨 (무한 루프)

### while Loop 동작 방식

1. 조건을 평가함
2. 조건이 True이면 루프 내의 코드 블록을 실행
3. 코드 블록 실행 후 다시 조건을 평가
4. 조건이 False가 될 때까지 반복

### The break Statement

- `break` 문을 사용하면 `while` 조건이 참이어도 루프를 중지할 수 있음
- `break`를 만나면 루프가 즉시 종료되고 루프 다음 코드로 이동
- 무한 루프를 안전하게 종료하는 데 유용함

### The continue Statement

- `continue` 문을 사용하면 현재 반복을 중지하고 다음 반복을 계속할 수 있음
- `continue`를 만나면 루프의 나머지 코드를 건너뛰고 조건 검사로 돌아감
- 특정 조건에서 일부 코드를 건너뛰고 싶을 때 유용함

### The else Statement

- `else` 문을 사용하면 조건이 더 이상 참이 아닐 때 코드 블록을 한 번 실행할 수 있음
- `while` 루프가 정상적으로 완료되었을 때 (break로 중단되지 않았을 때) 실행됨
- `break`로 루프를 중단한 경우 `else` 블록은 실행되지 않음


## Python 코드 실습

### While Loop 기본


````python
# While Loop 기본 - i가 6보다 작은 동안 출력
i = 1
while i < 6:
    print(i)
    i += 1
````

````python
# break 문 - i가 3일 때 루프 종료
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
````

### The continue Statement


````python
# continue 문 - i가 3일 때 현재 반복 건너뛰기
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
````

### The else Statement


````python
# else 문 - 조건이 False가 되었을 때 실행
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")
````

````python
# else 문과 break - break로 종료되면 else 실행 안 됨
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
else:
    print("This will not be printed because loop was broken")
````

````python
# 카운트다운 예제
countdown = 3
while countdown > 0:
    print(countdown)
    countdown -= 1
print("Happy New Year!!")
````

````python
# 조건이 처음부터 False인 경우 - 루프가 실행되지 않음
i = 10
while i < 5:
    print("This will never be printed")
    i += 1
print("Loop ended without execution")
````

## Java와의 비교

### 문법 차이

**Python:**
```python
# Python While Loop
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")

# break 문
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1

# continue 문
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

**Java:**
```java
// Java While Loop
int i = 1;
while (i < 6) {
    System.out.println(i);
    i++;
}

// break 문
int i = 1;
while (i < 6) {
    System.out.println(i);
    if (i == 3) {
        break;
    }
    i++;
}

// continue 문
int i = 0;
while (i < 6) {
    i++;
    if (i == 3) {
        continue;
    }
    System.out.println(i);
}

// Do/While Loop (Python에는 없음)
int i = 0;
do {
    System.out.println(i);
    i++;
} while (i < 5);
```

### 개념적 차이

- **While Loop 기본 구조**:
  - Python: 들여쓰기로 블록 정의. 조건 뒤에 콜론(`:`) 필요
  - Java: 중괄호 `{}`로 블록 정의. 조건을 괄호 `()`로 감싸야 함

- **While-Else 구조**:
  - Python: `while-else` 구조 지원. 루프가 정상적으로 완료되면 `else` 블록 실행
  - Java: `while-else` 구조 없음. `while` 루프에 `else`를 사용할 수 없음

- **Do-While Loop**:
  - Python: `do-while` 루프 없음. 최소 한 번 실행이 필요하면 조건을 조정하거나 다른 방법 사용
  - Java: `do-while` 루프 지원. 코드 블록을 먼저 실행한 후 조건 검사. 최소 한 번은 실행됨

- **Break Statement**:
  - Python: `break`로 루프 즉시 종료. `while-else`에서 `break`로 종료되면 `else` 블록 실행 안 됨
  - Java: `break`로 루프 즉시 종료. 동작 방식은 동일

- **Continue Statement**:
  - Python: `continue`로 현재 반복 건너뛰고 다음 반복 계속
  - Java: `continue`로 현재 반복 건너뛰고 다음 반복 계속. 동작 방식은 동일

- **무한 루프 방지**:
  - Python: 조건 변수를 증가시키는 것을 잊으면 무한 루프 발생
  - Java: 동일. 조건 변수를 증가시키는 것을 잊으면 무한 루프 발생

- **조건이 처음부터 False인 경우**:
  - Python: 루프가 전혀 실행되지 않음
  - Java: `while` 루프는 동일하게 실행되지 않음. `do-while`은 최소 한 번 실행됨


## 정리

### 핵심 내용

1. **While Loop**: 조건이 참인 동안 문 집합을 반복 실행
2. **Break Statement**: 조건이 참이어도 루프를 즉시 종료
3. **Continue Statement**: 현재 반복을 건너뛰고 다음 반복 계속
4. **Else Statement**: 루프가 정상적으로 완료되었을 때 (break로 중단되지 않았을 때) 실행
5. **무한 루프 방지**: 조건에서 사용하는 변수를 증가시키는 것을 잊지 말아야 함
6. **조건이 처음부터 False**: 루프가 전혀 실행되지 않음
7. **Python에는 do-while 없음**: 최소 한 번 실행이 필요하면 조건을 조정하거나 다른 방법 사용

### 느낀 점

- Python의 `while-else` 구조가 유용함. 루프가 정상 완료되었는지 확인할 수 있음.
- `break`와 `continue`가 Java와 동일하게 작동해서 이해하기 쉬움.
- `do-while`이 없어서 아쉽지만, 필요하면 조건을 조정하여 해결 가능.
- 들여쓰기로 블록을 정의하는 것이 깔끔하고 읽기 쉬움.
- 무한 루프를 방지하기 위해 항상 조건 변수를 증가시키는 것을 주의해야 함.
