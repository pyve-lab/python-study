## 학습 내용

### Python For Loops

- `for` 루프는 시퀀스(리스트, 튜플, 딕셔너리, 세트, 문자열)를 반복하는 데 사용됨
- 다른 프로그래밍 언어의 `for` 키워드와는 다르게, 객체 지향 프로그래밍 언어에서 찾을 수 있는 iterator 메서드처럼 작동함
- `for` 루프를 사용하면 리스트, 튜플, 세트 등의 각 항목에 대해 문 집합을 한 번씩 실행할 수 있음
- **중요**: `for` 루프는 사전에 인덱싱 변수를 설정할 필요가 없음

### For Loop 동작 방식

1. 시퀀스의 첫 번째 항목을 가져옴
2. 항목에 대해 코드 블록을 실행
3. 시퀀스의 다음 항목으로 이동
4. 시퀀스의 모든 항목을 처리할 때까지 반복

### Looping Through a String

- 문자열도 반복 가능한 객체임
- 문자열은 문자 시퀀스를 포함하므로 각 문자를 순회할 수 있음

### The break Statement

- `break` 문을 사용하면 모든 항목을 반복하기 전에 루프를 중지할 수 있음
- `break`를 만나면 루프가 즉시 종료됨

### The continue Statement

- `continue` 문을 사용하면 루프의 현재 반복을 중지하고 다음 반복을 계속할 수 있음
- `continue`를 만나면 루프의 나머지 코드를 건너뛰고 다음 항목으로 이동

### The range() Function

- 지정된 횟수만큼 코드 집합을 반복하려면 `range()` 함수를 사용할 수 있음
- `range()` 함수는 숫자 시퀀스를 반환함
- 기본적으로 0부터 시작하고 1씩 증가하며 지정된 숫자에서 끝남 (지정된 숫자는 포함하지 않음)
- `range(6)`은 0부터 5까지의 값 (6은 포함하지 않음)
- 시작 값을 지정할 수 있음: `range(2, 6)`은 2부터 5까지 (6은 포함하지 않음)
- 증가 값을 지정할 수 있음: `range(2, 30, 3)`은 2부터 29까지 3씩 증가

### Else in For Loop

- `for` 루프의 `else` 키워드는 루프가 완료되었을 때 실행할 코드 블록을 지정함
- **중요**: `break` 문으로 루프가 중단되면 `else` 블록은 실행되지 않음
- 루프가 정상적으로 완료되었을 때만 `else` 블록이 실행됨

### Nested Loops

- 중첩 루프는 루프 안의 루프임
- "내부 루프"는 "외부 루프"의 각 반복에 대해 한 번씩 실행됨

### The pass Statement

- `for` 루프는 비어 있을 수 없지만, 어떤 이유로 내용이 없는 `for` 루프가 있다면 `pass` 문을 넣어 오류를 피할 수 있음
- `pass` 문은 null 연산으로, 실행될 때 아무 일도 일어나지 않음


## Python 코드 실습


### For Loop 기본


````python
# For Loop 기본 - 리스트 순회
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
````

### Looping Through a String


````python
# 문자열 순회 - "banana"의 각 문자 출력
for x in "banana":
    print(x)
````

### The break Statement


````python
# break 문 - "banana"일 때 루프 종료
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
````

````python
# break 문 - print 전에 break (banana는 출력 안 됨)
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
````

### The continue Statement


````python
# continue 문 - "banana"는 출력하지 않고 다음 반복 계속
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
````

### The range() Function


````python
# range() 함수 기본 - 0부터 5까지 (6은 포함하지 않음)
for x in range(6):
    print(x)
````

````python
# range() 함수 - 시작 값 지정 (2부터 5까지)
for x in range(2, 6):
    print(x)
````

````python
# range() 함수 - 증가 값 지정 (2부터 29까지 3씩 증가)
for x in range(2, 30, 3):
    print(x)
````

````python
# range() 함수 - 역순 (5부터 1까지)
for x in range(5, 0, -1):
    print(x)
````

### Else in For Loop


````python
# else 문 - 루프가 정상적으로 완료되었을 때 실행
for x in range(6):
    print(x)
else:
    print("Finally finished!")
````

````python
# else 문과 break - break로 종료되면 else 실행 안 됨
for x in range(6):
    if x == 3:
        break
    print(x)
else:
    print("This will not be printed because loop was broken")
````

### Nested Loops


````python
# 중첩 루프 - 각 형용사에 대해 모든 과일 출력
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]
for x in adj:
    for y in fruits:
        print(x, y)
````

````python
# 중첩 루프 - 구구단 (2단부터 4단까지)
for i in range(2, 5):
    for j in range(1, 10):
        print(f"{i} x {j} = {i * j}")
    print()  # 빈 줄
````

### The pass Statement


````python
# pass 문 - 빈 for 루프
for x in [0, 1, 2]:
    pass
````

## Java와의 비교

### 문법 차이

**Python:**
```python
# Python For Loop - 시퀀스 순회
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)

# range() 함수 사용
for x in range(6):
    print(x)

# range() 함수 - 시작, 끝, 증가 값 지정
for x in range(2, 30, 3):
    print(x)

# for-else 구조
for x in range(6):
    print(x)
else:
    print("Finally finished!")

# 중첩 루프
for x in adj:
    for y in fruits:
        print(x, y)
```

**Java:**
```java
// Java For Loop - 전통적인 형태
for (int i = 0; i < 5; i++) {
    System.out.println(i);
}

// For-Each Loop (Python의 for와 유사)
String[] fruits = {"apple", "banana", "cherry"};
for (String x : fruits) {
    System.out.println(x);
}

// 중첩 루프
for (int i = 1; i <= 2; i++) {
    for (int j = 1; j <= 3; j++) {
        System.out.println(i + " " + j);
    }
}
```

### 개념적 차이

- **For Loop 기본 구조**:
  - Python: `for x in sequence:` 형태로 시퀀스를 직접 순회. 인덱스 변수 불필요
  - Java: 전통적인 `for (init; condition; increment)` 형태와 `for-each` 루프 두 가지 지원

- **시퀀스 순회**:
  - Python: `for x in sequence:` 형태로 리스트, 튜플, 딕셔너리, 세트, 문자열 등 모든 시퀀스 순회 가능
  - Java: `for-each` 루프 (`for (type var : array)`)로 배열이나 컬렉션 순회. 전통적인 for 루프는 인덱스 기반

- **range() 함수 vs 전통적인 for 루프**:
  - Python: `range()` 함수로 숫자 시퀀스 생성. `for x in range(6):` 형태
  - Java: 전통적인 for 루프로 숫자 반복. `for (int i = 0; i < 6; i++)` 형태

- **For-Else 구조**:
  - Python: `for-else` 구조 지원. 루프가 정상적으로 완료되면 `else` 블록 실행
  - Java: `for-else` 구조 없음. `for` 루프에 `else`를 사용할 수 없음

- **Break Statement**:
  - Python: `break`로 루프 즉시 종료. `for-else`에서 `break`로 종료되면 `else` 블록 실행 안 됨
  - Java: `break`로 루프 즉시 종료. 동작 방식은 동일

- **Continue Statement**:
  - Python: `continue`로 현재 반복 건너뛰고 다음 반복 계속
  - Java: `continue`로 현재 반복 건너뛰고 다음 반복 계속. 동작 방식은 동일

- **Nested Loops**:
  - Python: 중첩 루프 지원. 들여쓰기로 블록 정의
  - Java: 중첩 루프 지원. 중괄호로 블록 정의. 동작 방식은 동일

- **문자열 순회**:
  - Python: 문자열을 직접 순회 가능 (`for x in "banana":`)
  - Java: 문자열을 문자 배열로 변환하거나 `charAt()` 사용. 또는 `for-each`로 문자 배열 순회

- **인덱스 접근**:
  - Python: `enumerate()` 함수로 인덱스와 값 동시 접근 가능
  - Java: 전통적인 for 루프에서 인덱스 직접 사용. `for-each`에서는 인덱스 접근 어려움


## 정리

### 핵심 내용

1. **For Loop**: 시퀀스(리스트, 튜플, 딕셔너리, 세트, 문자열)를 반복하는 데 사용
2. **시퀀스 순회**: `for x in sequence:` 형태로 시퀀스를 직접 순회. 인덱스 변수 불필요
3. **문자열 순회**: 문자열도 반복 가능한 객체이므로 각 문자를 순회할 수 있음
4. **Break Statement**: 모든 항목을 반복하기 전에 루프를 중지
5. **Continue Statement**: 현재 반복을 건너뛰고 다음 반복 계속
6. **range() 함수**: 지정된 횟수만큼 반복. `range(6)`, `range(2, 6)`, `range(2, 30, 3)` 형태
7. **Else in For Loop**: 루프가 정상적으로 완료되었을 때 (break로 중단되지 않았을 때) 실행
8. **Nested Loops**: 루프 안의 루프. 내부 루프는 외부 루프의 각 반복에 대해 실행
9. **Pass Statement**: 빈 for 루프를 위한 플레이스홀더

### 느낀 점

- Python의 `for` 루프가 Java의 `for-each`와 유사하지만 더 일반적이고 강력함.
- `range()` 함수가 Java의 전통적인 for 루프와 유사한 역할을 하지만 더 간결함.
- `for-else` 구조가 유용함. 루프가 정상 완료되었는지 확인할 수 있음.
- 시퀀스를 직접 순회할 수 있어서 인덱스 관리가 필요 없어 편리함.
- 문자열도 직접 순회할 수 있어서 유용함.
- 중첩 루프가 테이블, 행렬, 다차원 데이터 구조 작업에 유용함.
