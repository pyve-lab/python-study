# 파이썬 리스트 (Lists): 내장 메서드 요약

파이썬은 리스트에 사용할 수 있는 일련의 **내장 메서드(Built-in Methods)**를 제공합니다. 이 메서드들은 리스트 항목을 추가, 제거, 검색, 정렬하는 등 다양한 작업을 수행할 수 있도록 합니다.

## 리스트 주요 메서드 (List Methods)

| 메서드 | 설명 (Description) 
| :--- | :--- |
| **`append()`** | 리스트의 맨 끝에 요소를 추가합니다. 
| **`clear()`** | 리스트에서 모든 요소를 제거합니다 (빈 리스트로 만듭니다). 
| **`copy()`** | 리스트의 사본을 반환합니다. 
| **`count()`** | 지정된 값을 가진 요소의 개수를 반환합니다. 
| **`extend()`** | 리스트 (또는 다른 반복 가능한 객체)의 요소를 현재 리스트의 끝에 추가합니다. 
| **`index()`** | 지정된 값을 가진 첫 번째 요소의 인덱스를 반환합니다. 
| **`insert()`** | 지정된 위치에 요소를 삽입합니다. 
| **`pop()`** | 지정된 위치의 요소를 제거하고 그 요소를 반환합니다. (인덱스 미지정 시 마지막 요소 제거) 
| **`remove()`** | 지정된 값과 일치하는 첫 번째 항목을 제거합니다. 
| **`reverse()`** | 리스트 요소의 순서를 역순으로 뒤집습니다. 
| **`sort()`** | 리스트를 정렬합니다. 

## 추가 학습: 검색 메서드 예시 (`count()`와 `index()`)

메서드 목록에는 포함되어 있지만, 이전 파일에서 직접 다루지 않았던 주요 검색 메서드인 `count()`와 `index()`의 사용법입니다.

```python
thislist = ["apple", "banana", "cherry", "apple", "kiwi"]

# 1. count(): 지정된 값의 개수 반환
count_apple = thislist.count("apple")
print(f"1. 'apple'의 개수: {count_apple}개")

# 2. index(): 지정된 값의 첫 번째 인덱스 반환
index_banana = thislist.index("banana")
print(f"2. 'banana'의 첫 번째 인덱스: {index_banana}")

# 참고: index() 메서드는 찾는 값이 리스트에 없으면 ValueError를 발생시킵니다.
```
