# 파이썬 리스트 (Lists): 항목 값 변경 및 삽입

## 1. 항목 값 변경 (Change Item Value)

특정 항목의 값을 변경하려면, 해당 항목의 **인덱스 번호**를 참조합니다.

```python
# 두 번째 항목("banana", 인덱스 1)을 "blackcurrant"로 변경:
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

## 2. 항목 값 범위 변경 (Change a Range of Item Values)

특정 범위 내 항목의 값을 변경하려면, **새로운 값들의 리스트**를 정의하고 **변경할 인덱스 범위**를 참조합니다.

```python
# "banana"(인덱스 1)와 "cherry"(인덱스 2)를 "blackcurrant"와 "watermelon"으로 변경
# 교체 항목 수와 삽입 항목 수가 같으므로 리스트 길이는 유지됩니다.
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
```

### 삽입 항목 > 교체 항목 (리스트 길이 증가)

교체할 항목보다 **더 많은** 항목을 삽입하면, 새 항목들은 지정된 위치에 삽입되고 나머지 항목들은 그에 따라 밀려납니다.

* **참고:** 삽입된 항목의 수가 교체된 항목의 수와 일치하지 않으면 리스트의 **길이가 변경**됩니다.

### 삽입 항목 < 교체 항목 (리스트 길이 감소)

교체할 항목보다 **더 적은** 항목을 삽입하면, 새 항목들은 지정된 위치에 삽입되고 나머지 항목들은 그에 따라 당겨집니다.

```python
# 1. 삽입 항목 > 교체 항목 (길이 증가):
# 두 번째 항목("banana", 인덱스 1)을 두 개의 새 값으로 교체 (길이 +1 증가)
list_increase = ["apple", "banana", "cherry"]
list_increase[1:2] = ["blackcurrant", "watermelon"]
print(f"1. 길이 증가 예시: {list_increase}")


# 2. 삽입 항목 < 교체 항목 (길이 감소):
# 두 번째("banana", 인덱스 1)와 세 번째("cherry", 인덱스 2) 항목을 하나의 값으로 교체 (길이 -1 감소)
list_decrease = ["apple", "banana", "cherry"]
list_decrease[1:3] = ["watermelon"]
print(f"2. 길이 감소 예시: {list_decrease}")
```

## 3. 항목 삽입 (Insert Items)

기존 항목을 교체하지 않고 **새로운 리스트 항목을 삽입**하려면 `insert()` 메서드를 사용할 수 있습니다.

`insert()` 메서드는 **지정된 인덱스**에 항목을 삽입합니다.

```python
# "watermelon"을 세 번째 항목(인덱스 2)으로 삽입:
thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(thislist)

# 결과적으로 리스트는 4개의 항목을 포함하게 됩니다.
```
