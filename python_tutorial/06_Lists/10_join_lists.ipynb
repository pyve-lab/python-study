# 파이썬 리스트 (Lists): 리스트 결합 (Joining Lists)

## 파이썬에서는 두 개 이상의 리스트를 **결합(Join)**하거나 **연결(Concatenate)**하는 여러 가지 방법이 있습니다.

## 1. '+' 연산자를 이용한 결합 (Concatenation)

리스트를 결합하는 가장 쉬운 방법 중 하나는 **`+` 연산자**를 사용하는 것입니다.

* 이 방법은 두 리스트의 요소를 모두 포함하는 **새로운 리스트**를 생성하여 반환합니다.

```python
# 두 리스트 결합 예시:
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(f"'+' 연산자를 사용한 결합 (list3): {list3}")
```

## 2. for 반복문과 append()를 이용한 결합

다른 방법은 `for` 반복문을 사용하여 `list2`의 모든 항목을 **`list1`에 하나씩 덧붙여(append)** 결합하는 것입니다.

* 이 방법은 `list1`을 직접 변경합니다.

```python
# list2의 항목들을 list1에 덧붙이기:
list1_for = ["a", "b", "c"]
list2_for = [1, 2, 3]

for x in list2_for:
  list1_for.append(x)

print(f"for/append를 사용한 결합 (list1_for 변경): {list1_for}")
```

## 3. extend() 메서드를 이용한 결합

가장 효율적이고 리스트 결합의 용도로 설계된 방법은 **`extend()`** 메서드를 사용하는 것입니다.

* 이 메서드는 다른 리스트의 요소를 현재 리스트의 **끝**에 추가합니다.
* `for` 반복문과 `append()`를 사용하는 것과 동일하게 **원본 리스트를 직접 변경**합니다.
