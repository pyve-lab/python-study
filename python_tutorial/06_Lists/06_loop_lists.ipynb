# 파이썬 리스트 (Lists): 반복문 사용 (Loop Lists)

리스트 항목을 순회하는 방법은 **for 반복문**, **인덱스를 이용한 for 반복문**, **while 반복문**, 그리고 **리스트 컴프리헨션(List Comprehension)**를 사용하는 등 다양합니다.

## 1. for 반복문을 이용한 순회 (Loop Through a List)

가장 일반적인 방법으로, `for` 반복문을 사용하여 리스트의 항목을 하나씩 순회할 수 있습니다.

```python
# 리스트의 모든 항목을 하나씩 출력:
thislist = ["apple", "banana", "cherry"]
for x in thislist:
  print(x)
```

## 2. 인덱스 번호를 이용한 순회 (Loop Through the Index Numbers)

`len()` 함수와 `range()` 함수를 사용하여 인덱스 번호를 참조하며 리스트 항목을 순회할 수도 있습니다.

* `len()`: 리스트의 길이를 구합니다.
* `range()`: 0부터 (리스트 길이 - 1)까지의 숫자를 생성합니다. (예시의 iterable은 `[0, 1, 2]`가 됩니다.)

```python
# 인덱스 번호를 참조하여 모든 항목 출력:
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
  print(thislist[i])
```

## 3. while 반복문을 이용한 순회 (Using a While Loop)

`while` 반복문을 사용하여 리스트 항목을 순회할 수도 있습니다.

1.  `len()` 함수로 리스트 길이를 구합니다.
2.  인덱스 변수 `i`를 0에서 시작하여 리스트를 순회합니다.
3.  반복 후에는 인덱스 변수를 **1씩 증가**시키는 것을 잊지 않아야 합니다.

```python
# while 반복문을 사용하여 인덱스 번호로 모든 항목 출력:
thislist = ["apple", "banana", "cherry"]
i = 0
while i < len(thislist):
  print(thislist[i])
  i = i + 1
```

## 4. 리스트 컴프리헨션 이용한 순회 (Looping Using List Comprehension)

**리스트 컴프리헨션(List Comprehension)**는 리스트를 순회하는 가장 짧은 구문을 제공합니다.

```python
# 리스트의 모든 항목을 출력하는 짧은 for 반복문 형태:
thislist = ["apple", "banana", "cherry"]
[print(x) for x in thislist]
```

```python
# 'a'가 포함된 항목만 모아 새 리스트를 생성하고 변수에 할당:
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x] # print()를 사용하지 않음
print(newlist) # 새 리스트의 내용을 출력
```
