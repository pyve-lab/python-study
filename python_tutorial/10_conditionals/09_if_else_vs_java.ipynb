# 파이썬 If 문과 자바 If 문 비교

파이썬과 자바는 모두 프로그램 흐름을 제어하기 위해 조건문(`if` 문)을 사용하지만, 각 언어의 디자인 철학에 따라 **코드 블록을 정의하는 문법**에서 가장 큰 차이를 보입니다.

---

## 1. 조건문 구조 및 문법 비교

| 구분 | 파이썬 (`if`/`elif`/`else`) | 자바 (`if`/`else if`/`else`) |
| :--- | :--- | :--- |
| **다중 조건 키워드** | **`elif`** (Else If의 축약어, 간결함) | **`else if`** (두 단어 사용) |
| **코드 블록 정의** | **들여쓰기(Indentation)** 사용 | **중괄호 `{}`** 사용 (스코프 정의) |
| **조건식 괄호** | 조건식에 **괄호 `()` 사용 불필요** | 조건식 주위에 **괄호 `()` 사용 필수** |
| **조건 끝 표시** | **콜론 `:`** 사용 | 중괄호로 대체 (콜론 불필요) |

````python
# 파이썬 If-Elif-Else 예시 (들여쓰기 기반)
score = 85

if score >= 90:
    print("A Grade")
elif score >= 80:
    print("B Grade") # 이 블록 실행
else:
    print("C Grade or lower")
````
