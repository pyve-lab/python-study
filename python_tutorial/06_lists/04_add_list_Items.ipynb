# 파이썬 리스트 (Lists): 항목 추가 (Adding Items)

## 1. 항목 추가 (Append Items)

리스트의 **맨 끝**에 항목을 추가하려면, `append()` 메서드를 사용합니다.

```python
# append() 메서드를 사용하여 항목 추가:
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)

# 결과적으로 리스트는 4개의 항목을 포함하게 됩니다.
```

## 2. 항목 삽입 (Insert Items)

지정된 인덱스 위치에 리스트 항목을 삽입하려면, `insert()` 메서드를 사용합니다.

* `insert(index, item)` 형식으로, 항목을 **지정된 인덱스**에 삽입합니다.

```python
# 두 번째 위치(인덱스 1)에 "orange" 삽입:
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)

# 결과적으로 리스트는 4개의 항목을 포함하게 됩니다.
```

## 3. 리스트 확장 (Extend List)

**다른 리스트의 요소**들을 현재 리스트의 **끝**에 추가하려면 `extend()` 메서드를 사용합니다.

```python
# tropical 리스트의 요소들을 thislist에 추가:
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)

# 요소들은 리스트의 맨 끝에 추가됩니다.
```

### 모든 반복 가능한 객체 추가 (Add Any Iterable)

`extend()` 메서드는 리스트뿐만 아니라, **모든 반복 가능한 객체(Iterable)**의 요소들(튜플, 세트, 딕셔너리 등)을 현재 리스트에 추가할 수 있습니다.

```python
# 튜플의 요소들을 리스트에 추가:
thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)
```
