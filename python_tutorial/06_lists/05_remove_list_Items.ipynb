# 파이썬 리스트 (Lists): 항목 제거 (Removing Items)

리스트에서 항목을 제거하는 방법은 **값으로 제거**하거나 **인덱스로 제거**하는 두 가지 주요 방법이 있습니다.

## 1. 특정 값 제거 (Remove Specified Item)

`remove()` 메서드는 리스트에서 **지정된 값**을 가진 첫 번째 항목을 제거합니다.

```python
# "banana" 항목 제거:
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)

# 중복 값 제거: remove()는 처음 발견된 항목 하나만 제거합니다.
thislist_duplicate = ["apple", "banana", "cherry", "banana", "kiwi"]
thislist_duplicate.remove("banana")
print(f"중복 제거 후: {thislist_duplicate}")
```

## 2. 특정 인덱스 제거 (Remove Specified Index)

### A. pop() 메서드 사용

`pop()` 메서드는 **지정된 인덱스**의 항목을 제거하고, 그 값을 반환합니다.

* 인덱스를 지정하지 않으면, `pop()` 메서드는 **마지막 항목**을 제거합니다.

```python
# 인덱스 1 (두 번째 항목) 제거:
list_pop_index = ["apple", "banana", "cherry"]
list_pop_index.pop(1)
print(f"인덱스 제거: {list_pop_index}")

# 인덱스를 지정하지 않고 마지막 항목 제거:
list_pop_last = ["apple", "banana", "cherry"]
list_pop_last.pop()
print(f"마지막 항목 제거: {list_pop_last}")
```

### B. del 키워드 사용

`del` 키워드는 **지정된 인덱스**의 항목을 제거합니다.

```python
# 인덱스 0 (첫 번째 항목) 제거:
thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
```

## 3. 리스트 전체 삭제 또는 내용 비우기

### A. 리스트 전체 삭제 (`del`)

`del` 키워드는 특정 항목뿐만 아니라 **리스트 변수 자체**를 완전히 삭제할 수도 있습니다.
(삭제 후에는 해당 변수를 참조하면 `NameError`가 발생합니다.)

### B. 리스트 내용 비우기 (`clear()`)

`clear()` 메서드는 리스트의 **내용만 모두 비웁니다**. 리스트 변수 자체는 그대로 남아있지만, 항목이 없는 **빈 리스트**(`[]`)가 됩니다.

```python
# 1. 리스트 전체 삭제 (del)
list_delete = ["apple", "banana", "cherry"]
del list_delete
# print(list_delete) # 이 코드를 실행하면 NameError가 발생합니다.


# 2. 리스트 내용 비우기 (clear())
list_clear = ["apple", "banana", "cherry"]
list_clear.clear()
print(f"내용 비우기: {list_clear}")
```
