# 파이썬 리스트 (Lists): 항목 접근 및 인덱싱

## 항목 접근 (Access Items)

리스트 항목은 인덱스(색인)를 가집니다. 인덱스 번호를 참조하여 항목에 접근할 수 있습니다.

* **참고:** 리스트의 첫 번째 항목은 인덱스 **0**을 가집니다.

```python
# 리스트에서 두 번째 항목(인덱스 1) 출력:
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```

## 음수 인덱싱 (Negative Indexing)

음수 인덱싱은 **리스트의 끝에서부터** 항목을 찾는 것을 의미합니다.

* **-1**은 **마지막 항목**을 참조합니다.
* **-2**는 **뒤에서 두 번째 항목**을 참조합니다.

```python
# 리스트의 마지막 항목(-1) 출력:
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

## 인덱스 범위 지정 (Range of Indexes)

시작 지점과 끝 지점을 지정하여 인덱스 범위를 지정할 수 있습니다.

* 범위를 지정하면 지정된 항목을 포함하는 **새로운 리스트**가 반환됩니다.
* **`[시작 인덱스: 끝 인덱스]`** 형식입니다.
* **참고:** 검색은 **시작 인덱스 (포함)**에서 시작하여 **끝 인덱스 (포함하지 않음)**에서 끝납니다.

```python
# 인덱스 2부터 인덱스 5 (포함하지 않음)까지의 항목 반환 (cherry, orange, kiwi)
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

```python
# 1. 시작 값 생략: 범위는 리스트의 처음부터 시작합니다.
# 처음부터 인덱스 4 (kiwi를 포함하지 않음)까지의 항목 반환
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])


# 2. 끝 값 생략: 범위는 리스트의 끝까지 계속됩니다.
# "cherry"(인덱스 2)부터 끝까지 항목 반환
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

## 음수 인덱스 범위 (Range of Negative Indexes)

리스트의 끝에서부터 검색을 시작하려면 음수 인덱스를 사용하여 범위를 지정할 수 있습니다.

```python
# 뒤에서 네 번째 항목("orange", -4)부터 뒤에서 첫 번째 항목("mango", -1)을 포함하지 않고 반환
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

## 항목 존재 여부 확인 (Check if Item Exists)

특정 항목이 리스트에 존재하는지 확인하려면 **`in`** 키워드를 사용합니다.
(이전에 학습한 멤버십 연산자를 사용합니다.)

```python
# "apple"이 리스트에 있는지 확인
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
```
