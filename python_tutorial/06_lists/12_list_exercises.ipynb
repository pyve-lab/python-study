# 파이썬 리스트 (Lists): 연습 문제


지금까지 리스트의 핵심 기능(접근, 변경, 추가, 제거, 컴프리헨션, 정렬 등)에 대해 학습했습니다. 

문제는 **객관식**과 **빈칸 채우기** 두 가지 유형으로 구성되어 있습니다.

```python
## 연습 문제 1: 항목 접근 및 변경 (Access & Change)

# 문제: 주어진 리스트에서 "banana" 항목을 "orange"로 변경하고,
# 리스트의 마지막 항목을 출력하는 코드로 올바른 것은 무엇입니까?

fruits = ["apple", "banana", "cherry", "kiwi"]

# 정답 코드:
fruits[1] = "orange"
print(fruits[-1])

# 결과: "kiwi"가 출력됩니다.
```

```python
## 연습 문제 2: 항목 추가 및 제거 (Add & Remove)

# 문제: 아래 코드가 ['lemon', 'apple', 'banana', 'cherry', 'grape']를 출력하도록
# 빈칸에 알맞은 메서드와 인덱스를 채우세요.

fruits = ["apple", "banana", "cherry"]

# 1. 리스트의 맨 앞에 "lemon" 삽입:
fruits.insert(0, "lemon")

# 2. 리스트의 맨 끝에 "grape" 추가:
fruits.append("grape")

print(fruits)

# 결과: ['lemon', 'apple', 'banana', 'cherry', 'grape']가 출력됩니다.
```

```python
## 연습 문제 3: 리스트 컴프리헨션 (List Comprehension)

# 문제: 다음 리스트 컴프리헨션의 실행 결과로 올바른 것은 무엇입니까?
# numbers = [1, 2, 3, 4, 5, 6]
# new_list = [x * 2 for x in numbers if x > 3]
# (정답: A. [8, 10, 12])

numbers = [1, 2, 3, 4, 5, 6]
new_list = [x * 2 for x in numbers if x > 3]

print(new_list)

# 결과: [8, 10, 12]가 출력됩니다. (4*2, 5*2, 6*2)
```

```python
## 연습 문제 4: 정렬 (Sorting)

# 문제: 아래 리스트를 대소문자를 구분하지 않고 알파벳 순서대로 정렬하려면
# 빈칸에 알맞은 key 인자를 채우세요.

words = ["Apple", "banana", "Cherry"]
words.sort(key = str.lower)

print(words)

# 결과: 대소문자 구분 없이 소문자 순으로 정렬됩니다. (예: ['Apple', 'banana', 'Cherry'] 또는 이와 유사한 순서)
```

```python
## 연습 문제 5: 복사 및 결합 (Copy & Join)

# 문제: 두 리스트 list1과 list2를 결합하여 ['a', 'b', 1, 2]를 만들고,
# 동시에 list1의 내용이 원본 그대로 유지되도록 하는 코드는 무엇입니까?
# (정답: B)

list1 = ["a", "b"]
list2 = [1, 2]

# B. 정답 코드: '+' 연산자는 새 리스트를 생성하여 원본을 유지합니다.
list3 = list1 + list2

print(list3)
print(f"list1 원본 유지 확인: {list1}")

# 결과: ['a', 'b', 1, 2]가 출력됩니다. list1은 ['a', 'b']로 유지됩니다.
```
