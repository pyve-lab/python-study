# 파이썬 리스트 (Lists) 기본

## 리스트 개요

리스트는 여러 항목을 하나의 변수에 저장하는 데 사용됩니다.

리스트는 파이썬의 **내장 자료형** 중 하나로, 데이터를 모아서 저장하는 데 사용되는 4가지 유형(List, Tuple, Set, Dictionary) 중 하나입니다. 이 네 가지 자료형은 각각 고유한 특징과 사용법을 가집니다.

리스트는 **대괄호 `[]`**를 사용하여 생성됩니다.

```python
# 리스트 생성 예시:
thislist = ["apple", "banana", "cherry"]
print(thislist)

# 예시 변수:
mylist = ["apple", "banana", "cherry"]
```

## 리스트 항목의 주요 특징

리스트 항목은 **순서가 있으며(Ordered)**, **변경 가능하고(Changeable)**, **중복 값(Allow Duplicates)**을 허용합니다.

### 1. 순서가 있음 (Ordered)

* 리스트가 순서가 있다는 것은 항목들이 정의된 순서를 가지며, 그 순서가 변하지 않음을 의미합니다.
* 항목들은 인덱스(색인)를 가지며, 첫 번째 항목은 인덱스 **`[0]`**, 두 번째 항목은 `[1]`을 가집니다.
* 새 항목을 리스트에 추가하면, 새 항목들은 리스트의 **끝**에 배치됩니다.
* **참고:** 리스트의 순서를 변경하는 일부 메서드(예: `.sort()`)가 있지만, 일반적으로 항목의 순서는 유지됩니다.

### 2. 변경 가능함 (Changeable / Mutable)

* 리스트는 **가변적(Mutable)**입니다. 즉, 리스트가 생성된 후에도 항목을 변경하거나, 추가하거나, 제거할 수 있습니다.

### 3. 중복 값 허용 (Allow Duplicates)

* 리스트 항목은 인덱스를 가지므로, 같은 값을 가진 항목들을 가질 수 있습니다.

```python
# 리스트는 중복 값을 허용하는 예시:
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```
