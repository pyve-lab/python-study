# 파이썬 자료구조: 세트 항목 접근 (Access Set Items)

## 항목 접근 (Access Items)

세트(Set)는 **순서가 없고(Unordered)** **인덱스가 없기(Unindexed)** 때문에, 리스트나 튜플처럼 인덱스(`[i]`)나 키를 사용하여 개별 항목에 직접 접근할 수 없습니다.

대신, 세트의 항목에 접근하는 방법은 주로 두 가지입니다.
1. `for` 루프를 사용하여 항목을 **순회(Loop)**합니다.
2. `in` 키워드를 사용하여 특정 값이 세트에 **존재하는지 확인**합니다.

## 1. `for` 루프를 사용한 순회

세트의 모든 항목을 순회하며 값을 출력하려면 `for` 루프를 사용합니다.

**참고:** 세트는 순서가 없으므로, 항목이 출력되는 순서는 실행할 때마다 다를 수 있습니다.

```python
# for 루프를 사용하여 세트 항목 순회 및 값 출력
thisset = {"apple", "banana", "cherry"}

for x in thisset:
  print(x)
```

## 2. `in` 키워드를 사용한 항목 존재 확인

특정 값이 세트 안에 존재하는지 확인하려면 **`in` 키워드**를 사용합니다. 결과는 `True` 또는 `False`로 반환됩니다.

```python
# "banana"가 세트에 있는지 확인
thisset = {"apple", "banana", "cherry"}

# "banana" in thisset
print(f"'banana' 존재 여부: {'banana' in thisset}")
```

```python
# "banana"가 세트에 없는지 확인
thisset = {"apple", "banana", "cherry"}

# "banana" not in thisset
print(f"'banana' 부재 여부: {'banana' not in thisset}")
```

## 3. 항목 변경 (Change Items) 불가

세트는 생성된 후에는 **항목의 값을 변경할 수 없습니다.**

하지만, 다음 챕터에서 배우겠지만, 항목을 **추가(add)**하거나 **제거(remove)**하여 세트 자체를 수정할 수는 있습니다.
