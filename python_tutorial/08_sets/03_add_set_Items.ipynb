# 파이썬 자료구조: 세트 항목 추가 (Add Set Items)

세트는 생성된 후에는 **기존 항목의 값을 변경할 수 없습니다.** 하지만, 필요에 따라 **새로운 항목을 추가**하거나, **다른 세트/컬렉션의 항목을 병합**할 수 있습니다.

세트에 항목을 추가하는 주요 방법은 두 가지 메서드입니다.
1.  **`add()`**: 세트에 **하나의 항목**을 추가합니다.
2.  **`update()`**: 현재 세트에 **다른 세트나 모든 순회 가능한 객체**의 항목을 병합합니다.

## 1. `add()` 메서드를 사용한 항목 추가

**`add()`** 메서드는 세트에 **하나의 항목**을 추가할 때 사용합니다.

**참고:** 세트는 중복을 허용하지 않으므로, 이미 존재하는 항목을 `add()` 해도 아무런 변화가 없습니다.

```python
# add() 메서드를 사용하여 "orange" 항목 하나를 추가
thisset = {"apple", "banana", "cherry"}

thisset.add("orange")

print(f"add() 후 세트: {thisset}")
```

## 2. `update()` 메서드를 사용한 세트 병합

**`update()`** 메서드는 **다른 세트**의 항목들을 현재 세트에 추가(병합)할 때 사용합니다.

```python
# update() 메서드를 사용하여 tropical 세트의 항목을 thisset에 병합
thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}

thisset.update(tropical)

print(f"update() 후 세트: {thisset}")
```

## 3. `update()` 메서드와 순회 가능한 객체

`update()` 메서드의 인수로 들어가는 객체는 반드시 세트일 필요는 없으며, **튜플, 리스트, 딕셔너리 등 모든 순회 가능한 객체(iterable object)**의 항목을 현재 세트에 병합할 수 있습니다.

**참고:** 리스트나 튜플의 항목이 세트에 추가될 때, **중복되는 값은 자동으로 제거**됩니다.

```python
# update() 메서드를 사용하여 리스트의 항목을 세트에 병합
thisset = {"apple", "banana", "cherry"}
mylist = ["kiwi", "orange"]

thisset.update(mylist)

print(f"리스트 병합 후 세트: {thisset}")
```
