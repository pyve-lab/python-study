# 파이썬 자료구조: 세트 항목 순회 (Loop Sets)

세트는 **순서가 없는(Unordered)** 컬렉션이므로, 튜플이나 리스트처럼 인덱스(`len()`과 `range()`)를 사용하여 항목을 순회할 수 없습니다.

세트의 항목을 순회하는 유일하고 가장 간단한 방법은 **`for` 루프**를 사용하는 것입니다.

## 1. `for` 루프를 사용한 항목 직접 순회
