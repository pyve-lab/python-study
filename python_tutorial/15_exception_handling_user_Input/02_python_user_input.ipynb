# Python 사용자 입력(User Input)

## 1. 기본 입력 받기: `input()` 함수

파이썬은 사용자의 입력을 받을 수 있는 **`input()`** 함수를 제공합니다. 프로그램 실행은 `input()` 함수를 만났을 때 멈추며, 사용자가 입력을 제공할 때까지 기다립니다.

```python
# 예시 1: 기본 입력 받기
print("Enter your name:")
name = input()
print(f"Hello {name}")
```

## 2. 프롬프트(Prompt) 사용하기

`input()` 함수는 **프롬프트 메시지**를 인수로 전달받아 입력 요청과 메시지를 같은 줄에 표시할 수 있습니다.

```python
# 예시 2: 프롬프트 메시지 사용
name = input("Enter your name: ")
print(f"Hello {name}")
```

## 3. 여러 개의 입력 받기

`input()` 함수를 여러 번 사용하여 필요한 만큼 입력을 받을 수 있습니다.

```python
# 예시 3: 여러 개의 입력 받기
name = input("Enter your name: ")
print(f"Hello {name}")
fav1 = input("What is your favorite animal: ")
fav2 = input("What is your favorite color: ")
fav3 = input("What is your favorite number: ")

print(f"Do you want a {fav2} {fav1} with {fav3} legs?")
```

## 4. 숫자 입력 처리 및 검증

### 4.1. 입력은 항상 문자열(String)

사용자로부터 받은 모든 입력은 숫자를 입력했더라도 **문자열(string)**로 처리됩니다. 계산을 수행하려면 **`float()`** 또는 **`int()`** 함수를 사용하여 입력값을 숫자로 **변환**해야 합니다.

```python
# 예시 4: 문자열을 숫자로 변환
import math

x = input("Enter a number: ")

# 숫자의 제곱근을 찾기 위해 float() 함수로 변환합니다.
y = math.sqrt(float(x))

print(f"The square root of {x} is {y}")
```

### 4.2. 입력 검증 (Validate Input)

유효성 검증은 좋은 관행입니다. 사용자가 숫자가 아닌 문자를 입력하면 에러가 발생하므로, **`try...except`** 구문과 **`while`** 루프를 사용하여 유효한 숫자를 받을 때까지 반복할 수 있습니다.

```python
# 예시 5: 숫자를 받을 때까지 반복 (입력 검증)
y = True
while y == True:
    x = input("Enter a number:")
    try:
        # 입력값을 숫자로 변환 시도
        x = float(x)
        # 에러가 발생하지 않으면 루프 종료
        y = False
    except:
        # 에러 발생 시 (숫자가 아닐 때) 사용자에게 재시도 메시지 출력
        print("Wrong input, please try again.")

print("Thank you!")
```
