## `while` Loop (While 반복문)

`while` 루프를 사용하면 **조건(`condition`)이 참(`True`)인 동안** 일련의 명령어들을 계속해서 실행할 수 있습니다.

### 기본 구조와 필수 요소

`while` 루프는 실행 전에 **인덱싱 변수(`indexing variable`)**를 정의해야 합니다.

**⚠️ Note (중요):** 반복문이 영원히 계속되는 **무한 루프(Infinite Loop)**를 방지하기 위해, 루프 내에서 **인덱스 변수를 반드시 증가/감소**시켜야 합니다. 

```python
# 예제 1: 기본 while 루프 (i가 6보다 작은 동안 i를 출력)

i = 1 # 인덱스 변수 초기화
while i < 6: # 조건: i가 6보다 작은 동안
  print(i)
  i += 1 # i를 증가시켜야 조건이 언젠가 False가 됨
  
```

## The `break` Statement (`break` 구문)

`break` 구문을 사용하면 `while` 조건이 여전히 참(`True`)일지라도 **즉시 반복문을 멈출(`stop the loop`)** 수 있습니다.

```python
# 예제 2: break 문을 사용하여 i가 3일 때 루프를 종료

i = 1
while i < 6:
  print(i)
  if i == 3:
    # i가 3일 때 조건과 상관없이 루프 탈출
    break
  i += 1
```

## The `continue` Statement (`continue` 구문)

`continue` 구문은 현재 반복(`current iteration`)을 **중지시키고**, 루프의 **다음 반복으로 계속 진행**하도록 합니다. 즉, `continue` 이후의 코드는 실행되지 않습니다.

**⚠️ 주의:** `continue`를 사용하기 전에 인덱스 변수(`i += 1`)를 변경해야 무한 루프에 빠지는 것을 방지할 수 있습니다.

```python
# 예제 3: continue 문을 사용하여 i가 3일 때 출력 건너뛰기

i = 0
while i < 6:
  i += 1 # continue 전에 i를 증가시켜야 함
  if i == 3:
    # i가 3일 때 print(i)를 건너뛰고 바로 다음 루프 시작
    continue
  print(i) # 3은 출력되지 않음
```

## The `else` Statement (`else` 구문)

`else` 구문은 `while` **조건이 더 이상 참이 아닐 때** (`when the condition no longer is true`), **단 한 번** 실행되는 코드 블록을 실행할 수 있습니다.

**ℹ️ 참고:** 만약 반복문이 `break` 문에 의해 강제로 종료되었다면, `else` 블록은 실행되지 않습니다.

```python
# 예제 4: else 문을 사용하여 조건이 거짓이 되었을 때 메시지 출력

i = 1
while i < 6:
  print(i)
  i += 1
else:
  # while 루프가 정상적으로 완료되면 실행됨
  print("i is no longer less than 6")
```
