# 파이썬 튜플: 튜플 결합 및 곱하기 (Join & Multiply Tuples)

튜플은 **`+` 연산자**를 사용하여 두 개 이상의 튜플을 연결(Concatenate)하거나, **`*` 연산자**를 사용하여 내용을 반복(Multiply)할 수 있습니다.

## 1. 튜플 결합 (`+` 연산자 사용)

두 개 이상의 튜플을 결합(Join)하려면 **`+` 연산자**를 사용하면 됩니다.

* **결과:** 이 연산은 기존 튜플을 변경하지 않고, 두 튜플의 항목을 모두 포함하는 **새로운 튜플**을 반환합니다.

```python
# 두 튜플 결합 예시:
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)

# 새로운 튜플 tuple3 생성
tuple3 = tuple1 + tuple2
print(f"결합된 튜플: {tuple3}")
```

## 2. 튜플 곱하기 (`*` 연산자 사용)

튜플의 내용을 지정된 횟수만큼 **반복**하여 새로운 튜플을 만들려면 **`*` 연산자**를 사용합니다.

* **결과:** 튜플의 항목이 지정된 횟수만큼 반복되는 새로운 튜플이 생성됩니다.

```python
# fruits 튜플을 2배로 곱하기:
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(f"반복된 튜플: {mytuple}")
```
