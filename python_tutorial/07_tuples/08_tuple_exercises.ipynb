## 1. 튜플 기본 (Tuples Basic)

### 문제 1: 튜플 판별
다음 변수 중 튜플로 인식되는 것은 무엇인지 코드로 확인하고, 튜플이 아닌 경우 그 이유를 설명하세요.

```python
# 문제 1 정답 코드
A = ("banana")
B = ("apple",)

print(f"A의 타입: {type(A)}")
print(f"B의 타입: {type(B)}")

# 풀이: A는 쉼표가 없으므로 문자열(str)로 인식되고, B는 쉼표(,)가 있으므로 튜플(tuple)로 인식됩니다.
```

## 2. 항목 접근 (Access Tuples)

### 문제 2: 음수 인덱싱
다음 튜플에서 **"kiwi"** 항목을 **음수 인덱싱**을 사용하여 출력하는 코드를 작성하세요.

```python
# 문제 2 정답 코드
fruits = ("apple", "banana", "cherry", "orange", "kiwi", "melon")

# 'kiwi'는 끝에서 두 번째 항목이므로 인덱스는 -2입니다.
print(fruits[-2])
```

## 3. 항목 업데이트 (Update Tuples)

### 문제 3: 불변성 극복
튜플의 불변성을 극복하고, 다음 튜플의 **"cherry"** 항목을 **"grape"**로 변경하는 코드를 작성하세요.

```python
# 문제 3 정답 코드
thistuple = ("apple", "banana", "cherry")

# 1. 튜플을 리스트로 변환
temp_list = list(thistuple)

# 2. 리스트에서 항목 변경 (인덱스 2)
temp_list[2] = "grape"

# 3. 리스트를 다시 튜플로 변환
thistuple = tuple(temp_list)

print(thistuple)
```

## 4. 언패킹 (Unpack Tuples)

### 문제 4: 애스터리스크 * 사용
다음 튜플을 `(first, *middle, last)`로 언패킹했을 때, `middle` 변수에 할당되는 값이 **리스트** 형태로 출력되도록 코드를 작성하세요.

```python
# 문제 4 정답 코드
data = ("A", "B", "C", "D", "E")

# 첫 번째와 마지막을 제외한 나머지를 *middle에 리스트로 모읍니다.
(first, *middle, last) = data

print(f"middle 변수: {middle}")
```

## 5. 항목 순회 (Loop Tuples)

### 문제 5: 인덱스 순회
다음 튜플 항목을 **인덱스 번호**를 사용하여 순회하고 모든 항목을 출력하는 `for` 루프 코드를 작성하세요.

```python
# 문제 5 정답 코드
colors = ("red", "green", "blue")

# len()과 range()를 사용하여 인덱스 0, 1, 2를 생성합니다.
for i in range(len(colors)):
  print(colors[i])
```

## 6. 튜플 결합 (Join Tuples)

### 문제 6: + 연산자
다음 두 튜플을 **`+` 연산자**를 사용하여 결합한 `combined_data` 튜플을 만들고 출력하세요.

```python

# 문제 6 정답 코드
info = ("name", "age")
values = ("John", 30)

# + 연산자를 사용하여 두 튜플을 결합합니다.
combined_data = info + values

print(combined_data)
```

## 7. 튜플 메서드 (Tuple Methods)

### 문제 7: count() 사용
다음 튜플에서 **숫자 7**이 **몇 번** 등장하는지 확인하는 코드를 작성하세요.

```python
# 문제 7 정답 코드
numbers = (7, 2, 5, 7, 3, 7, 1)

# count() 메서드를 사용하여 특정 값의 등장 횟수를 계산합니다.
count_seven = numbers.count(7)

print(f"숫자 7의 등장 횟수: {count_seven}")
```
