# 파이썬 튜플: 항목 순회 (Loop Tuples)

튜플 항목은 리스트와 마찬가지로 **`for` 루프**나 **`while` 루프**를 사용하여 순회할 수 있습니다.

## 1. `for` 루프를 사용한 항목 직접 순회

가장 간단하고 일반적인 방법입니다. 튜플의 항목을 하나씩 가져와 값을 출력하거나 처리할 수 있습니다.

```python
# for x in tuple: 항목을 순서대로 순회하며 값 출력
thistuple = ("apple", "banana", "cherry")

for x in thistuple:
  print(x)
```

## 2. `for` 루프를 사용한 인덱스 번호 순회

항목 자체 대신 **인덱스 번호**를 참조하여 순회할 수도 있습니다.
이 방법은 항목의 값뿐만 아니라 **순서(인덱스)**가 필요할 때 유용합니다.

* `range()` 함수와 `len()` 함수를 함께 사용하여 순회 가능한 인덱스 범위를 만듭니다.

```python
# len()으로 길이(3)를 구하고, range(3)으로 인덱스 범위(0, 1, 2)를 생성하여 순회
thistuple = ("apple", "banana", "cherry")

for i in range(len(thistuple)):
  print(f"인덱스 {i}: {thistuple[i]}")
```

## 3. `while` 루프 사용

`while` 루프를 사용하여 인덱스 번호를 통해 튜플 항목을 순회할 수도 있습니다.

1.  `i = 0`으로 시작 인덱스를 설정합니다.
2.  `while i < len(튜플)` 조건으로 튜플의 길이만큼 반복합니다.
3.  반복할 때마다 `i`를 1씩 증가시켜 다음 인덱스로 이동해야 합니다.

```python
# while 루프를 사용하여 모든 인덱스 번호를 순회하며 항목 출력
thistuple = ("apple", "banana", "cherry")
i = 0

while i < len(thistuple):
  print(thistuple[i])
  # 다음 항목으로 이동하기 위해 i를 1 증가
  i = i + 1
```
