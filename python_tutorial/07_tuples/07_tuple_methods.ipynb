# 파이썬 튜플: 튜플 메서드 (Tuple Methods)

튜플은 변경 불가능한(Immutable) 자료형이므로, 리스트(List)처럼 항목을 추가하거나 제거, 변경하는 메서드(예: `append()`, `remove()`)를 가지지 않습니다.

파이썬은 튜플에서 사용할 수 있는 단 **두 가지**의 내장 메서드만 제공합니다.

## 1. `count()` 메서드

### 기능
`count()` 메서드는 튜플 내에서 지정된 **특정 값이 나타나는 횟수**를 반환합니다.

### 형식
`tuple.count(value)`

```python
# 'apple'이 튜플에 몇 번 나타나는지 횟수 계산
thistuple = (1, 5, 7, "apple", 5, "apple", 9, 3, "apple")

# 'apple'의 등장 횟수
count_apple = thistuple.count("apple")
print(f"튜플: {thistuple}")
print(f"'apple'의 등장 횟수: {count_apple}")

# 숫자 5의 등장 횟수
count_five = thistuple.count(5)
print(f"숫자 5의 등장 횟수: {count_five}")
```

## 2. `index()` 메서드

### 기능
`index()` 메서드는 튜플 내에서 지정된 **특정 값**을 검색하고, 해당 값이 **처음 발견된 위치의 인덱스**를 반환합니다.

**주의:** 찾는 값이 튜플 내에 존재하지 않으면 **에러(ValueError)**를 발생시킵니다.

### 형식
`tuple.index(value)`

```python
# 'cherry'가 처음 나타나는 위치의 인덱스 검색
thistuple = ("apple", "banana", "cherry", "orange", "cherry", "kiwi")

# 'cherry'의 인덱스 (가장 먼저 나오는 인덱스 2를 반환)
index_cherry = thistuple.index("cherry")
print(f"튜플: {thistuple}")
print(f"'cherry'의 인덱스: {index_cherry}")

# 'banana'의 인덱스
index_banana = thistuple.index("banana")
print(f"'banana'의 인덱스: {index_banana}")

# 값이 없는 경우 (주석 처리): 주석 해제 후 실행 시 ValueError 발생
# thistuple.index("mango")
```
