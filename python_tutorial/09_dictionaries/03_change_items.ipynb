# 파이썬 딕셔너리 항목 수정 (Change Dictionary Items)

## 1. 개별 항목의 값 변경 (Change Values)

딕셔너리의 특정 항목 값(Value)은 해당 항목의 **키 이름(Key name)**을 참조하여 새 값을 할당하는 방식으로 변경할 수 있습니다.

### 1.1 대괄호 `[]`를 사용하여 변경
변경하려는 키 이름을 대괄호 `[]` 안에 넣고 `=` 연산자를 사용하여 새 값을 할당합니다.

```python
# 딕셔너리 정의
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# "year" 값을 2018로 변경
thisdict["year"] = 2018

print(f"값 변경 후 딕셔너리: {thisdict}")
```

## 2. 여러 항목 업데이트 (`update()`)

`.update()` 메서드는 주어진 인수(argument)의 항목들로 딕셔너리를 업데이트합니다.

* **용도:** 기존 키의 **값을 변경**하거나, 존재하지 않는 키를 사용하여 **새 항목을 추가**할 때 사용됩니다.
* **인수:** 인수는 반드시 **다른 딕셔너리**이거나 `key:value` 쌍을 가진 순회 가능한 객체(iterable object)여야 합니다.

```python
# .update() 메서드를 사용하여 값 변경
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# update()를 사용하여 "year" 값을 2020으로 변경
thisdict.update({"year": 2020})

print(f"update() 후 딕셔너리: {thisdict}")
```

```python
# .update()를 사용하여 새 항목 추가
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# "color" 항목을 새로 추가
thisdict.update({"color": "red"})

print(f"update()로 추가 후 딕셔너리: {thisdict}")
```
