# 파이썬 딕셔너리 항목 추가 (Add Dictionary Items)

## 1. 개별 항목 추가 (Adding Items)

딕셔너리에 새 항목을 추가하는 가장 기본적인 방법은 **새로운 인덱스 키(key)**를 사용하고 그 키에 값을 할당하는 것입니다.

* **방식:** `딕셔너리[새 키] = 새 값`

이 방식은 키가 이미 존재하면 **값을 수정**하고, 키가 존재하지 않으면 **새 항목을 추가**합니다.

```python
# 딕셔너리 정의
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# 새 키("color")와 값("red")을 할당하여 항목 추가
thisdict["color"] = "red"

print(f"항목 추가 후 딕셔너리: {thisdict}")
```

## 2. 여러 항목 또는 조건부 추가 (`update()`)

`.update()` 메서드는 주어진 인수의 항목들로 딕셔너리를 업데이트합니다. 이 메서드는 기존 키의 값을 수정하는 데 사용되지만, **항목이 존재하지 않으면 새 항목을 추가**하는 역할도 수행합니다.

* **용도:** 여러 항목을 동시에 추가하거나 수정할 때 유용합니다.
* **인수:** 인수는 반드시 **딕셔너리**이거나 `key:value` 쌍을 가진 순회 가능한 객체(iterable object)여야 합니다.

```python
# .update() 메서드를 사용하여 항목 추가
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# update()를 사용하여 "color" 항목을 새로 추가
thisdict.update({"color": "red"})

print(f"update()로 추가 후 딕셔너리: {thisdict}")
```

```python
# .update()로 여러 항목 추가 및 기존 항목 수정 동시 진행
my_car = {
    "brand": "BMW",
    "model": "X5",
    "year": 2010
}

# 새 항목 추가("engine") 및 기존 항목 수정("year")
my_car.update({"engine": "v8", "year": 2022})

print(f"update() 동시 진행 후 딕셔너리: {my_car}")
```
