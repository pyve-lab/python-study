# 파이썬 딕셔너리 항목 제거 (Remove Dictionary Items)

## 1. 딕셔너리 항목 제거 방법

딕셔너리에서 항목을 제거하는 데는 여러 메서드와 키워드가 사용됩니다.

### 1.1 `pop()` 메서드: 키를 지정하여 제거
`pop()` 메서드는 지정된 **키(key)**에 해당하는 항목을 제거합니다. 제거된 값은 반환됩니다.

```python
# pop() 메서드를 사용하여 "model" 항목 제거
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

removed_value = thisdict.pop("model")

print(f"제거된 값: {removed_value}")
print(f"pop() 후 딕셔너리: {thisdict}")
```

### 1.2 `popitem()` 메서드: 마지막 항목 제거
`popitem()` 메서드는 딕셔너리의 마지막에 삽입된 항목(키-값 쌍)을 제거합니다.

* **참고:** Python 3.7 이전 버전에서는 무작위 항목을 제거했습니다. 현재는 **마지막 항목**을 제거하고 해당 항목(튜플 형태)을 반환합니다.

```python
# popitem() 메서드를 사용하여 마지막 항목 제거
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

removed_item = thisdict.popitem()

print(f"제거된 항목: {removed_item}")
print(f"popitem() 후 딕셔너리: {thisdict}")
```

## 2. `del` 키워드를 사용한 제거

### 2.1 특정 키를 지정하여 항목 제거
`del` 키워드는 지정된 키에 해당하는 항목을 제거합니다.

```python
# del 키워드를 사용하여 "model" 항목 제거
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

del thisdict["model"]

print(f"del 후 딕셔너리: {thisdict}")
```

### 2.2 딕셔너리 전체 삭제
`del` 키워드는 딕셔너리 변수 자체를 메모리에서 완전히 삭제하는 데에도 사용될 수 있습니다.

```python
# del 키워드를 사용하여 딕셔너리 전체 삭제
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

del thisdict

# print(thisdict) # 이 코드를 실행하면 NameError가 발생합니다. (thisdict가 더 이상 존재하지 않으므로)
print("thisdict 변수가 메모리에서 삭제되었습니다. 다시 접근하면 NameError가 발생합니다.")
```

## 3. `clear()` 메서드를 사용한 내용 비우기

`clear()` 메서드는 딕셔너리 **변수 자체는 유지**하되, 딕셔너리 **내부의 모든 항목을 비워서** 빈 딕셔너리 `{}`로 만듭니다.

```python
# clear() 메서드를 사용하여 딕셔너리 내용 비우기
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

thisdict.clear()

print(f"clear() 후 딕셔너리: {thisdict}")
```
