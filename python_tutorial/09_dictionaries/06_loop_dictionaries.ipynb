# 파이썬 딕셔너리 반복 (Loop Dictionaries)

## 1. 딕셔너리 순회 기본

딕셔너리는 **`for` 루프**를 사용하여 순회할 수 있습니다. 딕셔너리 자체를 `for` 루프에 사용하면 기본적으로 **키(Key)**가 반환됩니다.

### 1.1 기본 순회: 키(Key) 반환

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# 기본 순회 시 'x'에는 키(Key)가 할당됨
print("--- 1. 기본 순회 (키) ---")
for x in thisdict:
  print(x)
```

## 2. 값(Value)을 순회하는 방법

### 2.1 키를 이용해 값 접근
기본 순회로 얻은 키(`x`)를 사용하여 딕셔너리 항목 값(`thisdict[x]`)에 접근하여 출력할 수 있습니다.

```python
# 키를 이용해 값 출력
print("--- 2.1 키를 이용한 값 순회 ---")
for x in thisdict:
  print(thisdict[x])
```

### 2.2 `values()` 메서드 사용
`.values()` 메서드를 루프에 직접 사용하여 **값(Value)**만 효율적으로 반환받을 수 있습니다.

```python
# values() 메서드를 이용한 값 순회
print("--- 2.2 values() 메서드 순회 ---")
for x in thisdict.values():
  print(x)
```

## 3. 키를 명시적으로 순회 (`keys()`)

`.keys()` 메서드를 루프에 명시적으로 사용하여 **키(Key)**만 반환받을 수도 있습니다. 이는 기본 순회와 결과는 같지만, 키를 순회한다는 의도를 명확히 할 때 사용됩니다.

```python
# keys() 메서드를 이용한 키 순회
print("--- 3. keys() 메서드 순회 ---")
for x in thisdict.keys():
  print(x)
```

## 4. 키와 값을 동시에 순회 (`items()`)

`.items()` 메서드는 루프에 가장 많이 사용되는 방법 중 하나로, 각 항목을 **(키, 값) 튜플**로 반환합니다. 루프 변수를 두 개(`x, y`)로 설정하여 키와 값을 동시에 가져올 수 있습니다.

```python
# items() 메서드를 이용한 키와 값 동시 순회
print("--- 4. items() 메서드 순회 (키, 값 동시) ---")
for x, y in thisdict.items():
  print(f"Key: {x}, Value: {y}")
```
