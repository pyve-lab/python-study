### 🐍 What is Python?  
Python은 매우 인기 있는 프로그래밍 언어입니다.  
Guido van Rossum이 만들었고, **1991년에 처음 출시**되었습니다.

### 🧾 Python의 활용 분야
- 웹 개발 (서버 사이드)
- 소프트웨어 개발
- 수학/과학 계산
- 시스템 스크립팅

---

### 🛠 What can Python do?
- 서버에서 웹 애플리케이션을 만들 수 있음  
- 다양한 소프트웨어와 연동하여 워크플로우 자동화 가능  
- 데이터베이스 시스템 연결, 파일 읽기/쓰기 가능  
- 빅데이터 처리 및 복잡한 수학 계산 수행  
- 프로토타입(시제품) 개발부터 실제 서비스 개발까지 모두 가능  

---

### ⭐ Why Python? (파이썬을 쓰는 이유)
- Windows, Mac, Linux, Raspberry Pi 등 다양한 플랫폼에서 실행 가능  
- 영어에 가까운 단순하고 직관적인 문법  
- 다른 언어보다 **더 적은 코드 줄 수**로 동일한 기능 구현 가능  
- 인터프리터 방식 → 코드를 작성하자마자 즉시 실행 가능 → **빠른 개발 속도**  
- 절차적(Procedural), 객체지향(OOP), 함수형(Function) 방식 모두 지원  

---

### 📌 Good to Know (알아두면 좋은 점)
- 최신 주요 버전은 **Python 3**이며, 본 튜토리얼에서도 Python 3 기준  
- 텍스트 에디터에서 코드를 작성할 수 있음  
- Thonny, PyCharm, NetBeans, Eclipse 같은 IDE에서 개발할 수도 있으며  
  특히 많은 Python 파일을 관리할 때 유용함  

---

### 🔤 Python Syntax vs 다른 언어
- **읽기 쉬운 문법** (영어 + 수학 스타일 영향)  
- 대부분의 언어가 ;(세미콜론)으로 명령을 끝내는 것과 달리  
  → Python은 **줄바꿈(New line)** 으로 명령을 완료  
- 들여쓰기(Indentation)를 사용해 범위(Scope)를 정의  
  → 반복문, 함수, 클래스 등  
- 다른 언어(C, Java, C++, JS 등)는 `{}`(중괄호)를 범위 지정에 사용함  

---

### ☕ What is Java?
Java는 강력하고 인기 있는 프로그래밍 언어로, **1995년에 출시**되었습니다.  
현재 Java는 Oracle이 관리하고 있으며, **30억 개 이상의 기기에서 Java가 실행**되고 있습니다.

### 🧾 Java의 활용 분야
- 모바일 앱(특히 안드로이드)
- 데스크톱 애플리케이션
- 웹 애플리케이션
- 웹 서버 / 애플리케이션 서버
- 게임 개발
- 데이터베이스 연동
- 그 외 수많은 분야에서 사용

---

### ⭐ Why Use Java? (자바를 쓰는 이유)
- 다양한 플랫폼(Windows, Mac, Linux, Raspberry Pi 등)에서 실행 가능  
- 세계에서 가장 인기 있는 프로그래밍 언어 중 하나  
- 현재 IT 취업 시장에서 지속적인 수요  
- 배우기 쉽고 사용하기 쉬움  
- 오픈소스이며 무료  
- 빠르고 안전하며 강력함  
- 수천만 명 규모의 대규모 커뮤니티  
- 명확한 구조를 제공하는 **객체지향 언어(OOP)**  
  → 코드 재사용성 증가 → 개발 비용 절감  
- C++ 또는 C#과 유사하여 다른 언어로의 전환이 용이  

---

### 📘 Python vs Java 간단 비교

| 항목 | Python | Java |
|------|--------|-------|
| 출시 | 1991 | 1995 |
| 실행 방식 | 인터프리터 | 컴파일 + JVM |
| 문법 | 간결, 영어 같음 | 엄격하고 구조적 |
| 범위(Scope) | 들여쓰기 | 중괄호 `{ }` |
| 속도 | 상대적으로 느림 | 상대적으로 빠름 |
| 스타일 | 절차적 + OOP + 함수형 | OOP 중심 |
| 사용 분야 | 데이터/AI/웹/자동화 | 대규모 서비스, 앱, 서버 |
| 난이도 | 쉬움 | 조금 더 엄격함 |
