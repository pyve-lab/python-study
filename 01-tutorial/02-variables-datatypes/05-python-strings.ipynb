## 학습 내용

### Strings 기본

- 작은따옴표(`'`) 또는 큰따옴표(`"`)로 문자열을 둘러쌈
- `'hello'`와 `"hello"`는 동일함
- 문자열 내부에 따옴표를 사용할 때는 다른 따옴표로 감싸면 됨
- 멀티라인 문자열은 세 개의 따옴표(`"""` 또는 `'''`) 사용

### Strings are Arrays

- Python의 문자열은 유니코드 문자 배열
- Python에는 문자(char) 타입이 없음. 단일 문자도 길이가 1인 문자열
- 대괄호 `[]`를 사용하여 문자열의 요소에 접근 가능
- `for` 루프로 문자열의 각 문자를 순회 가능

### String Length & Check

- `len()` 함수로 문자열 길이 확인
- `in` 키워드로 특정 문구나 문자가 문자열에 있는지 확인
- `not in` 키워드로 특정 문구나 문자가 문자열에 없는지 확인

### Slicing Strings

- 슬라이싱 문법으로 문자열의 일부를 반환 가능
- `[start:end]` 형식 (end는 포함되지 않음)
- 시작 인덱스 생략 시 처음부터, 끝 인덱스 생략 시 끝까지
- 음수 인덱스로 끝에서부터 접근 가능

### Modify Strings

- 문자열 메서드: `upper()`, `lower()`, `strip()`, `replace()`, `split()`
- 모든 문자열 메서드는 새로운 값을 반환 (원본 문자열은 변경되지 않음)

### String Concatenation

- `+` 연산자로 두 문자열을 결합

### Format Strings

- F-Strings (Python 3.6+): `f"문자열 {변수}"` 형식
- `format()` 메서드 사용 가능
- Placeholder에 변수, 연산, 함수, 포맷팅 수정자 포함 가능

### Escape Characters

- 백슬래시(`\`)로 특수 문자를 문자열에 삽입
- 주요 이스케이프 문자: `\'`, `\"`, `\\`, `\n`, `\t` 등

### String Methods

- Python은 다양한 내장 문자열 메서드 제공
- 주요 메서드: `capitalize()`, `count()`, `find()`, `index()`, `join()`, `startswith()`, `endswith()` 등


## Python 코드 실습


````python
# Strings 기본 - 작은따옴표와 큰따옴표
print("Hello")
print('Hello')

# 문자열 내부에 따옴표 사용
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')

# 변수에 문자열 할당
a = "Hello"
print(a)
````

````python
# 멀티라인 문자열 - 세 개의 큰따옴표
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)

# 세 개의 작은따옴표
b = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(b)
````

````python
# Strings are Arrays - 인덱싱 (첫 번째 문자의 위치는 0)
a = "Hello, World!"
print(a[1])  # 'e' 출력

# 문자열 순회
for x in "banana":
    print(x)
````

````python
# String Length - len() 함수
a = "Hello, World!"
print(len(a))  # 13 출력
````

````python
# Check String - in 키워드
txt = "The best things in life are free!"
print("free" in txt)  # True

# if 문에서 사용
if "free" in txt:
    print("Yes, 'free' is present.")

# not in 키워드
print("expensive" not in txt)  # True

if "expensive" not in txt:
    print("No, 'expensive' is NOT present.")
````

````python
# Slicing Strings - [start:end] (end는 포함되지 않음)
b = "Hello, World!"
print(b[2:5])  # 'llo' 출력 (인덱스 2부터 4까지)

# 시작 인덱스 생략 (처음부터)
print(b[:5])  # 'Hello' 출력

# 끝 인덱스 생략 (끝까지)
print(b[2:])  # 'llo, World!' 출력

# 음수 인덱스
print(b[-5:-2])  # 'orl' 출력 (끝에서 5번째부터 끝에서 2번째 전까지)
````

````python
# Modify Strings - upper(), lower()
a = "Hello, World!"
print(a.upper())  # 'HELLO, WORLD!'
print(a.lower())  # 'hello, world!'

# strip() - 앞뒤 공백 제거
a = " Hello, World! "
print(a.strip())  # 'Hello, World!'

# replace() - 문자열 치환
a = "Hello, World!"
print(a.replace("H", "J"))  # 'Jello, World!'

# split() - 구분자로 분리하여 리스트 반환
a = "Hello, World!"
print(a.split(","))  # ['Hello', ' World!']
````

````python
# String Concatenation - + 연산자
a = "Hello"
b = "World"
c = a + b
print(c)  # 'HelloWorld'

# 공백 추가
c = a + " " + b
print(c)  # 'Hello World'
````

````python
# Format Strings - F-Strings (Python 3.6+)
age = 36
txt = f"My name is John, I am {age}"
print(txt)  # 'My name is John, I am 36'

# Placeholder에 포맷팅 수정자
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)  # 'The price is 59.00 dollars'

# Placeholder에 연산 포함
txt = f"The price is {20 * 59} dollars"
print(txt)  # 'The price is 1180 dollars'
````

````python
# Escape Characters
# 큰따옴표 안에 큰따옴표 사용
txt = "We are the so-called \"Vikings\" from the north."
print(txt)

# 작은따옴표 안에 작은따옴표 사용
txt = 'It\'s alright.'
print(txt)

# 백슬래시
txt = "The character \\ is called backslash."
print(txt)

# 줄바꿈
txt = "Hello\nWorld"
print(txt)

# 탭
txt = "Hello\tWorld"
print(txt)
````

````python
# String Methods - 주요 메서드 예시
txt = "hello world"

# capitalize() - 첫 문자를 대문자로
print(txt.capitalize())  # 'Hello world'

# count() - 특정 값의 개수
print(txt.count('l'))  # 3

# find() - 특정 값의 위치 (없으면 -1)
print(txt.find('world'))  # 6

# index() - 특정 값의 위치 (없으면 에러)
print(txt.index('world'))  # 6

# startswith() - 특정 값으로 시작하는지
print(txt.startswith('hello'))  # True

# endswith() - 특정 값으로 끝나는지
print(txt.endswith('world'))  # True

# join() - 문자열로 이터러블 결합
myTuple = ("John", "Peter", "Vicky")
x = "#".join(myTuple)
print(x)  # 'John#Peter#Vicky'
````

## Java와의 비교

### 문자열 선언 및 기본 사용

**Python:**
```python
# 작은따옴표 또는 큰따옴표 사용 가능
a = "Hello"
b = 'Hello'

# 멀티라인 문자열
c = """Line 1
Line 2
Line 3"""

# 문자열은 배열처럼 접근
print(a[0])  # 'H'
for char in a:
    print(char)

# 문자열 길이
print(len(a))  # 5

# 문자열 포함 확인
print("ell" in a)  # True
```

**Java:**
```java
// 큰따옴표만 사용
String a = "Hello";

// 멀티라인 문자열 (Java 13+)
String c = """
    Line 1
    Line 2
    Line 3
    """;

// 문자열은 배열처럼 접근
System.out.println(a.charAt(0));  // 'H'
for (char ch : a.toCharArray()) {
    System.out.println(ch);
}

// 문자열 길이
System.out.println(a.length());  // 5

// 문자열 포함 확인
System.out.println(a.contains("ell"));  // true
```

### 문자열 메서드

**Python:**
```python
txt = "Hello World"

# 대소문자 변환
print(txt.upper())  # 'HELLO WORLD'
print(txt.lower())  # 'hello world'

# 공백 제거
txt2 = " Hello World "
print(txt2.strip())  # 'Hello World'

# 문자열 치환
print(txt.replace("H", "J"))  # 'Jello World'

# 문자열 분리
print(txt.split(" "))  # ['Hello', 'World']

# 문자열 찾기
print(txt.find("World"))  # 6
print(txt.index("World"))  # 6

# 시작/끝 확인
print(txt.startswith("Hello"))  # True
print(txt.endswith("World"))  # True
```

**Java:**
```java
String txt = "Hello World";

// 대소문자 변환
System.out.println(txt.toUpperCase());  // "HELLO WORLD"
System.out.println(txt.toLowerCase());  // "hello world"

// 공백 제거
String txt2 = " Hello World ";
System.out.println(txt2.trim());  // "Hello World"

// 문자열 치환
System.out.println(txt.replace("H", "J"));  // "Jello World"

// 문자열 분리 (배열 반환)
String[] parts = txt.split(" ");
System.out.println(Arrays.toString(parts));  // [Hello, World]

// 문자열 찾기
System.out.println(txt.indexOf("World"));  // 6

// 시작/끝 확인
System.out.println(txt.startsWith("Hello"));  // true
System.out.println(txt.endsWith("World"));  // true
```

### 문자열 결합 및 포맷팅

**Python:**
```python
# + 연산자로 결합
a = "Hello"
b = "World"
c = a + " " + b  # 'Hello World'

# F-Strings (Python 3.6+)
age = 36
txt = f"My name is John, I am {age}"  # 'My name is John, I am 36'

# format() 메서드
txt2 = "My name is John, I am {}".format(age)
```

**Java:**
```java
// + 연산자로 결합
String a = "Hello";
String b = "World";
String c = a + " " + b;  // "Hello World"

// concat() 메서드
String c2 = a.concat(" ").concat(b);

// String.format() 또는 printf 스타일
int age = 36;
String txt = String.format("My name is John, I am %d", age);

// 문자열과 숫자 결합 (자동 변환)
String txt2 = "My name is John, I am " + age;  // 자동으로 문자열로 변환
```

### 개념적 차이

- **따옴표**: 
  - Python: 작은따옴표(`'`)와 큰따옴표(`"`) 모두 사용 가능
  - Java: 큰따옴표(`"`)만 사용
- **문자 타입**: 
  - Python: 문자(char) 타입 없음. 단일 문자도 길이가 1인 문자열
  - Java: `char` 타입 존재. `String`은 객체
- **문자열 접근**: 
  - Python: `[인덱스]`로 직접 접근 가능
  - Java: `charAt(인덱스)` 메서드 사용
- **문자열 길이**: 
  - Python: `len()` 함수 사용
  - Java: `length()` 메서드 사용
- **문자열 포함 확인**: 
  - Python: `in`, `not in` 키워드 사용
  - Java: `contains()` 메서드 사용
- **슬라이싱**: 
  - Python: `[start:end]` 문법으로 슬라이싱 지원
  - Java: `substring(start, end)` 메서드 사용
- **문자열 포맷팅**: 
  - Python: F-Strings (가장 선호), `format()` 메서드
  - Java: `String.format()`, `printf` 스타일, 문자열 결합
- **문자열과 숫자 결합**: 
  - Python: F-Strings나 `str()` 변환 필요
  - Java: `+` 연산자로 자동 변환 (숫자가 문자열로 변환됨)
- **이스케이프 문자**: 
  - Python과 Java 모두 백슬래시(`\`) 사용. 주요 이스케이프 문자는 동일


## 정리

### 핵심 내용

1. Python의 문자열은 작은따옴표(`'`) 또는 큰따옴표(`"`)로 둘러쌀 수 있으며, 멀티라인 문자열은 세 개의 따옴표 사용
2. 문자열은 유니코드 문자 배열로, 인덱싱(`[인덱스]`)과 슬라이싱(`[start:end]`)으로 접근 가능
3. `len()` 함수로 문자열 길이 확인, `in`/`not in` 키워드로 포함 여부 확인
4. 문자열 메서드(`upper()`, `lower()`, `strip()`, `replace()`, `split()` 등)는 새로운 값을 반환 (원본 변경 없음)
5. `+` 연산자로 문자열 결합, F-Strings로 문자열 포맷팅 (Python 3.6+)
6. 이스케이프 문자(`\'`, `\"`, `\\`, `\n`, `\t` 등)로 특수 문자 삽입
7. Python에는 문자(char) 타입이 없어 단일 문자도 길이가 1인 문자열
8. Java와 달리 작은따옴표/큰따옴표 모두 사용 가능, F-Strings로 더 간편한 포맷팅

### 느낀 점

- Python의 문자열 슬라이싱 문법이 직관적이고 편리.
- F-Strings가 Java의 String.format보다 훨씬 읽기 쉽고 간결.
- `in`/`not in` 키워드가 `contains()` 메서드보다 더 자연스러움.
- 문자열 메서드가 모두 새로운 값을 반환해서 원본이 변경되지 않는 것이 안전.
