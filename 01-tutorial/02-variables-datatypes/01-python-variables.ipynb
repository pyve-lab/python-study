## 학습 내용

### Variables (변수)

변수는 데이터 값을 저장하는 컨테이너.

### Creating Variables

Python은 변수 선언 명령어가 없음. 값 할당 시 자동으로 변수 생성됨.

### 타입 변경

변수는 특정 타입으로 선언할 필요가 없으며, 할당된 후에도 타입을 변경할 수 있음.

### Casting

변수의 데이터 타입을 명시적으로 지정하려면 casting 사용.

### Get the Type

`type()` 함수로 변수의 데이터 타입 확인 가능.

### Variable Names

- 문자 또는 언더스코어(`_`)로 시작
- 숫자로 시작 불가
- 영문자, 숫자, 언더스코어만 사용 가능
- 하이픈(`-`) 사용 불가 (뺄셈 연산자로 해석됨)
- 대소문자 구분
- Python 키워드 사용 불가

### 언더스코어(`_`)로 시작하는 변수명

Python에서 언더스코어로 시작하는 변수는 관례적으로 특별한 의미를 가짐:
- `_variable`: 내부 사용을 나타내는 관례 (private처럼, 하지만 실제로는 접근 가능)
- `__variable`: Name mangling (클래스 내에서 사용)
- `__variable__`: 특수 메서드 (magic methods)

Python은 접근 제어자(private, public 등)가 없어서, 언더스코어로 "내부용"임을 표시하는 관례를 사용함.

### Multiple Values

- 여러 변수에 여러 값 할당: `x, y, z = "Orange", "Banana", "Cherry"`
- 여러 변수에 같은 값 할당: `x = y = z = "Orange"`
- Collection unpacking: 리스트나 튜플 등의 컬렉션에서 값을 개별 변수로 풀어내는 것. `x, y, z = fruits` (fruits의 첫 번째 값이 x에, 두 번째 값이 y에, 세 번째 값이 z에 할당됨)

### Output Variables

`print()` 함수로 변수 출력. 쉼표(`,`)로 여러 변수 출력 가능. `+` 연산자로 문자열 연결 가능.

### Global Variables

함수 외부에서 생성된 변수는 전역 변수. 
- 함수 내부에서 전역 변수를 **읽기만** 하면 `global` 키워드 불필요
- 함수 내부에서 전역 변수를 **변경**하려면 `global` 키워드 필요
- `global` 키워드 없이 함수 내부에서 같은 이름의 변수에 할당하면 **지역 변수**가 됨 (전역 변수는 변경되지 않음)

### Constants (상수)

Python은 Java의 `final`처럼 언어 레벨에서 상수를 강제하는 키워드가 없음. 대신 관례적으로 대문자로 변수명을 작성하여 상수임을 나타냄:
- `CONSTANT_NAME = value`: 관례적으로 상수로 사용 (실제로는 변경 가능하지만, 개발자들이 변경하지 않기로 약속)
- Java의 `final` 키워드처럼 컴파일/실행 시점에 변경을 막지는 않음


## Python 코드 실습


````python
# 변수 생성 (값 할당 시 자동 생성)
x = 5
y = "John"
print(x)
print(y)
````

````python
# 타입 변경 가능
x = 4       # x는 int 타입
x = "Sally" # x는 이제 str 타입
print(x)
````

````python
# Casting (타입 명시)
x = str(3)    # x는 '3'
y = int(3)    # y는 3
z = float(3)  # z는 3.0
print(x, y, z)
````

````python
# 언더스코어로 시작하는 변수명
_my_var = "John"  # 관례적으로 내부 사용을 나타냄
my_var = "John"   # 일반 변수
print(_my_var, my_var)
````

````python
# 타입 확인
x = 5
y = "John"
print(type(x))
print(type(y))
````

````python
# 여러 변수에 여러 값 할당
x, y, z = "Orange", "Banana", "Cherry"
print(x, y, z)
````

````python
# 여러 변수에 같은 값 할당
x = y = z = "Orange"
print(x, y, z)
````

````python
# Collection unpacking (컬렉션에서 값을 개별 변수로 풀어내기)
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits  # fruits의 첫 번째 값이 x에, 두 번째 값이 y에, 세 번째 값이 z에 할당
print(x, y, z)
````

````python
# 변수 출력
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)  # 쉼표로 구분
print(x + " " + y + " " + z)  # + 연산자로 연결
````

````python
# Global Variables
x = "awesome"

def myfunc():
    print("Python is " + x)

myfunc()
````

````python
# Global Variable vs Local Variable
# global 키워드 없이 함수 내부에서 같은 이름의 변수에 할당하면 지역 변수가 됨
x = 'awesome'

def myfunc():
    x = 'fantastic'  # 이 x는 지역 변수 (전역 변수 x와는 별개)

myfunc()
print('Python is ' + x)  # 전역 변수 x는 변경되지 않음
````

````python
# 상수 (관례적으로 대문자 사용)
PI = 3.14159
MINUTES_PER_HOUR = 60
BIRTH_YEAR = 1990

# 실제로는 변경 가능하지만, 관례적으로 변경하지 않음
print(PI, MINUTES_PER_HOUR, BIRTH_YEAR)
````

````python
# global 키워드로 전역 변수 변경
x = "awesome"

def myfunc():
    global x
    x = "fantastic"

myfunc()
print("Python is " + x)
````

## Java와의 비교

### 변수 선언 및 사용 차이

**Python:**
```python
# 값 할당 시 자동 생성, 타입 선언 불필요
x = 5
x = "Sally"  # 타입 변경 가능

# 여러 값 할당
x, y, z = "Orange", "Banana", "Cherry"
x = y = z = "Orange"

# 상수 (관례적으로 대문자 사용)
PI = 3.14159
# PI = 3.14  # 실제로는 변경 가능하지만, 관례적으로 변경하지 않음

# 변수 출력
print(x, y, z)  # 쉼표로 구분
```

**Java:**
```java
// 타입 선언 필수
int x = 5;
// x = "Sally";  // 에러: 타입 변경 불가

// 여러 변수 선언 (같은 타입만)
int x = 5, y = 6, z = 50;
x = y = z = 50;

// 상수 선언 (final 키워드)
final int MINUTES_PER_HOUR = 60;
// MINUTES_PER_HOUR = 70;  // 에러: final 변수는 변경 불가

// 클래스 레벨 변수 (전역 변수처럼 사용)
public class Main {
    static String x = "awesome";  // 클래스 변수 (static)
    
    public static void myMethod() {
        System.out.println("Java is " + x);  // 클래스 변수 사용
        x = "fantastic";  // 클래스 변수 변경 (static이므로 직접 변경 가능)
    }
    
    public static void main(String[] args) {
        myMethod();
        System.out.println("Java is " + x);  // 변경된 값 출력
    }
}

// 변수 출력
System.out.println(x + " " + y + " " + z);
```

### 개념적 차이

- **변수 선언**: Python은 값 할당 시 자동 생성, Java는 타입 선언 필수
- **타입 변경**: Python은 가능, Java는 불가능
- **여러 값 할당**: Python은 다양한 패턴 지원, Java는 제한적
- **상수**: 
  - Python: 언어 레벨의 상수 키워드 없음. 관례적으로 대문자 변수명 사용 (`CONSTANT_NAME`). 실제로는 변경 가능하지만 개발자들이 변경하지 않기로 약속
  - Java: `final` 키워드로 상수 선언. 컴파일 시점에 변경 시도하면 에러 발생
- **전역 변수**: 
  - Python: 함수 외부에서 생성된 변수는 전역 변수. 함수 내부에서 전역 변수 변경 시 `global` 키워드 필요
  - Java: 클래스 레벨 변수로 사용. `static` 키워드와 함께 사용하여 클래스 변수로 선언
- **변수 이름 규칙**: 
  - Python: 영문자, 숫자, 언더스코어(`_`)만 사용 가능. 하이픈(`-`) 사용 불가 (뺄셈 연산자로 해석)
  - Java: 영문자, 숫자, 언더스코어(`_`), 달러(`$`) 사용 가능. 하이픈(`-`) 사용 불가
- **언더스코어 사용**: 
  - Python: `_variable`로 시작하는 변수는 관례적으로 내부 사용을 나타냄 (접근 제어자가 없어서 관례 사용)
  - Java: 언더스코어로 시작 가능하지만 일반적으로 사용하지 않음. `private`, `protected`, `public` 등의 접근 제어자를 사용


## 정리

### 핵심 내용

1. Python은 값 할당 시 변수가 자동으로 생성됨
2. 변수 타입을 변경할 수 있음
3. 여러 변수에 여러 값 또는 같은 값을 한 줄에 할당 가능
4. `global` 키워드로 함수 내부에서 전역 변수 변경 가능
5. 상수는 관례적으로 대문자 변수명 사용 (Java의 `final`처럼 강제하지 않음)
6. Java와 달리 타입 선언이 필요 없고 더 유연함

### 느낀 점

- Java처럼 타입을 선언할 필요가 없어서 코드가 훨씬 간결.
- 변수 타입을 변경할 수 있어서 유연하지만, 실수할 가능성도 있음.
- 여러 값 할당 패턴이 다양해서 편리.
