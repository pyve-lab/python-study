## Java와의 비교

### 실행 방식 차이

**Python:**
```python
# hello.py
print("Hello, World!")

# 실행: python hello.py
```

**Java:**
```java
// Main.java
public class Main {
    public static void main(String[] args) {
        System.out.println("Hello World");
    }
}

// 컴파일: javac Main.java
// 실행: java Main
```

### 인터프리터 vs 컴파일

- **인터프리터 (Python)**: 소스 코드를 한 줄씩 읽어서 바로 실행. 실행 시점에 코드를 해석하여 동작.
- **컴파일 (Java)**: 소스 코드를 먼저 기계어나 바이트코드로 변환한 후 실행. 컴파일 단계를 거쳐 실행 파일 생성.

### 개념적 차이

- **실행 방식**: Python은 인터프리터로 직접 실행, Java는 컴파일 후 실행
- **파일 구조**: Python은 단순 스크립트, Java는 클래스와 파일명이 일치해야 함
- **명령어**: Python은 `python 파일명.py`, Java는 `javac 파일명.java` 후 `java 클래스명`
- **속도**: Python은 즉시 실행, Java는 컴파일 단계 필요


## 정리

### 핵심 내용

1. Python은 인터프리터 언어로 파일 작성 후 바로 실행 가능
2. `python 파일명.py` 명령으로 간단히 실행
3. Python Command Line으로 간단한 코드 테스트 가능

### 느낀 점

- Java처럼 컴파일 과정 없이 바로 실행할 수 있어서 빠르게 테스트 가능.
- 파일 구조가 Java보다 훨씬 간단.
