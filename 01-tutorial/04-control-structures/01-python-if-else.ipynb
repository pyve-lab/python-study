## 학습 내용

### Python Conditions and If Statements

Python은 수학에서 사용하는 일반적인 논리 조건을 지원:

- **Equals**: `a == b`
- **Not Equals**: `a != b`
- **Less than**: `a < b`
- **Less than or equal to**: `a <= b`
- **Greater than**: `a > b`
- **Greater than or equal to**: `a >= b`

이러한 조건들은 여러 방법으로 사용할 수 있으며, 가장 일반적으로 "if statements"와 루프에서 사용.

### If Statement 기본

- `if` 키워드를 사용하여 if 문 작성
- 조건이 True이면 if 블록 내의 코드가 실행됨
- 조건이 False이면 if 블록이 건너뛰어짐

### Indentation (들여쓰기)

- Python은 들여쓰기(줄 시작 부분의 공백)를 사용하여 코드의 범위를 정의
- 다른 프로그래밍 언어는 종종 중괄호를 사용하지만, Python은 들여쓰기를 사용
- 같은 코드 블록 내의 모든 문은 동일한 들여쓰기 양을 사용해야 함
- 공백 또는 탭을 사용할 수 있지만, 같은 블록 내에서는 일관되게 사용해야 함

### Multiple Statements in If Block

- if 블록 내에 여러 문을 포함할 수 있음
- 모든 문은 동일한 레벨에서 들여쓰기되어야 함

### Using Variables in Conditions

- Boolean 변수는 비교 연산자 없이 if 문에서 직접 사용 가능
- Python은 if 문에서 많은 타입의 값을 True 또는 False로 평가할 수 있음
- `0`, 빈 문자열(`""`), `None`, 빈 컬렉션은 False로 처리됨
- 그 외의 모든 것은 True로 처리됨 (양수, 음수, 비어있지 않은 문자열 등)

### Elif Statement

- `elif` 키워드는 "이전 조건이 참이 아니면, 이 조건을 시도해보세요"를 의미
- 여러 표현식을 확인하고 조건 중 하나가 True로 평가되면 코드 블록을 실행
- 여러 개의 elif 문을 사용할 수 있음
- Python은 순서대로 각 조건을 확인하고 첫 번째로 참인 조건을 실행
- **중요**: 첫 번째로 참인 조건만 실행됨. 여러 조건이 참이어도 Python은 첫 번째 일치하는 블록을 실행한 후 중지

### Else Statement

- `else` 키워드는 앞의 조건들에 의해 잡히지 않은 모든 것을 잡음
- if 조건(및 모든 elif 조건)이 False로 평가될 때 else 문이 실행됨
- else 문은 마지막에 와야 함. else 다음에 elif를 사용할 수 없음
- else는 기본 동작을 제공하며, if와 elif 문으로 다루지 않은 시나리오에 대한 "catch-all" 역할

### Shorthand If (한 줄 If)

- 실행할 문이 하나만 있는 경우, if 문과 같은 줄에 배치할 수 있음
- 조건 뒤에 콜론(`:`)이 여전히 필요함

### Shorthand If ... Else (Ternary Operator)

- if와 else에 각각 하나의 문이 있는 경우, 조건식(conditional expression)을 사용하여 한 줄에 배치할 수 있음
- 이것을 "ternary operator"라고도 함
- 구문: `variable = value_if_true if condition else value_if_false`
- 조건식을 연결할 수 있지만, 가독성을 위해 짧게 유지해야 함

### Logical Operators (논리 연산자)

논리 연산자는 조건문을 결합하는 데 사용됩니다. Python에는 세 가지 논리 연산자가 있습니다:

- **and**: 두 문이 모두 참이면 True 반환
- **or**: 문 중 하나가 참이면 True 반환
- **not**: 결과를 반전시킴. 결과가 참이면 False 반환

**연산자 우선순위**: Python은 `not`을 먼저 평가한 다음 `and`, 그 다음 `or`를 평가합니다.

### Nested If Statements

- if 문 안에 if 문을 가질 수 있음. 이것을 nested if statements라고 함
- 각 중첩 레벨은 더 깊은 의사 결정 레벨을 만듦
- 코드는 가장 바깥쪽 조건부터 안쪽으로 평가됨
- 너무 많은 레벨의 중첩은 코드를 읽기 어렵게 만들 수 있음
- 때로는 nested if 문을 `and`와 같은 논리 연산자로 단순화할 수 있음

### Pass Statement

- if 문은 비어 있을 수 없지만, 어떤 이유로 내용이 없는 if 문이 있다면 `pass` 문을 넣어 오류를 피할 수 있음
- `pass` 문은 null 연산으로, 실행될 때 아무 일도 일어나지 않음
- 플레이스홀더 역할을 하며, 코드 구조를 만들 때나 나중에 구현할 로직을 위한 자리 표시자로 유용함
- 주석과 달리 `pass`는 실제 문이며 실행됨 (아무 일도 하지 않지만)


## Python 코드 실습


### If Statement


````python
# If statement 기본
a = 33
b = 200
if b > a:
    print("b is greater than a")
````

````python
# 숫자가 양수인지 확인
number = 15
if number > 0:
    print("The number is positive")
````

````python
# If 블록 내 여러 문
age = 20
if age >= 18:
    print("You are an adult")
    print("You can vote")
    print("You have full legal rights")
````

````python
# Boolean 변수 사용
is_logged_in = True
if is_logged_in:
    print("Welcome back!")
````

````python
# 다양한 값의 Truthiness 확인
# False로 평가되는 값들
print(bool(0))        # False
print(bool(""))       # False
print(bool(None))     # False
print(bool([]))       # False
print(bool({}))       # False
print(bool(()))       # False

# True로 평가되는 값들
print(bool(5))        # True
print(bool(-3))       # True
print(bool("False"))  # True (비어있지 않은 문자열)
print(bool([1, 2]))   # True
````

### Elif Statement


````python
# Elif 기본
a = 33
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
````

````python
# 여러 Elif 문 - 점수 등급
score = 75
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
````

````python
# 연령대 분류
age = 25
if age < 13:
    print("You are a child")
elif age < 20:
    print("You are a teenager")
elif age < 65:
    print("You are an adult")
elif age >= 65:
    print("You are a senior")
````

````python
# 요일 확인
day = 3
if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
````

### Else Statement


````python
# If-Elif-Else 완전한 체인
a = 200
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
````

````python
# Else without Elif
a = 200
b = 33
if b > a:
    print("b is greater than a")
else:
    print("b is not greater than a")
````

````python
# 짝수/홀수 확인
number = 7
if number % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
````

````python
# 온도 분류기
temperature = 22
if temperature > 30:
    print("It's hot outside!")
elif temperature > 20:
    print("It's warm outside")
elif temperature > 10:
    print("It's cool outside")
else:
    print("It's cold outside!")
````

````python
# 사용자 입력 검증
username = "Emil"
if len(username) > 0:
    print(f"Welcome, {username}!")
else:
    print("Error: Username cannot be empty")
````

### Shorthand If


````python
# 한 줄 If 문
a = 5
b = 2
if a > b: print("a is greater than b")
````

````python
# 한 줄 If ... Else (Ternary Operator) - 값 출력
a = 2
b = 330
print("A") if a > b else print("B")
````

````python
# Ternary Operator로 값 할당
a = 10
b = 20
bigger = a if a > b else b
print("Bigger is", bigger)
````

````python
# 여러 조건을 한 줄에 연결
a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")
````

````python
# 두 수의 최댓값 찾기
x = 15
y = 20
max_value = x if x > y else y
print("Maximum value:", max_value)
````

````python
# 기본값 설정
username = ""
display_name = username if username else "Guest"
print("Welcome,", display_name)
````

### Logical Operators (논리 연산자)


````python
# AND 연산자
a = 200
b = 33
c = 500
if a > b and c > a:
    print("Both conditions are True")
````

````python
# OR 연산자
a = 200
b = 33
c = 500
if a > b or a > c:
    print("At least one of the conditions is True")
````

````python
# NOT 연산자
a = 33
b = 200
if not a > b:
    print("a is NOT greater than b")
````

````python
# 여러 논리 연산자 결합
age = 25
is_student = False
has_discount_code = True
if (age < 18 or age > 65) and not is_student or has_discount_code:
    print("Discount applies!")
````

````python
# 괄호를 사용한 명확한 조건
temperature = 25
is_raining = False
is_weekend = True
if (temperature > 20 and not is_raining) or is_weekend:
    print("Great day for outdoor activities!")
````

````python
# 사용자 인증 확인
username = "Tobias"
password = "secret123"
is_verified = True
if username and password and is_verified:
    print("Login successful")
else:
    print("Login failed")
````

````python
# 범위 확인
score = 85
if score >= 0 and score <= 100:
    print("Valid score")
else:
    print("Invalid score")
````

### Nested If Statements


````python
# Nested If 기본
x = 41
if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
    else:
        print("but not above 20.")
````

````python
# 여러 조건을 중첩으로 확인
age = 25
has_license = True
if age >= 18:
    if has_license:
        print("You can drive")
    else:
        print("You need a license")
else:
    print("You are too young to drive")
````

````python
# 세 단계 중첩
score = 85
attendance = 90
submitted = True
if score >= 60:
    if attendance >= 80:
        if submitted:
            print("Pass with good standing")
        else:
            print("Pass but missing assignment")
    else:
        print("Pass but low attendance")
else:
    print("Fail")
````

````python
# Nested If vs Logical Operators
# Nested If 방식
temperature = 25
is_sunny = True
if temperature > 20:
    if is_sunny:
        print("Perfect beach weather!")

# Logical Operator 방식 (더 간단)
if temperature > 20 and is_sunny:
    print("Perfect beach weather!")
````

````python
# 로그인 검증 (중첩 체크)
username = "Emil"
password = "python123"
is_active = True
if username:
    if password:
        if is_active:
            print("Login successful")
        else:
            print("Account is not active")
    else:
        print("Password required")
else:
    print("Username required")
````

````python
# 등급 계산 (중첩 로직)
score = 92
extra_credit = 5
if score >= 90:
    if extra_credit > 0:
        print("A+ grade")
    else:
        print("A grade")
elif score >= 80:
    print("B grade")
else:
    print("C grade or below")
````

### Pass Statement


````python
# Pass 문 기본
a = 33
b = 200
if b > a:
    pass
````

````python
# 개발 중 플레이스홀더
age = 16
if age < 18:
    pass  # TODO: Add underage logic later
else:
    print("Access granted")
````

````python
# Pass vs Comments
# 주석만 사용하면 IndentationError 발생
# score = 85
# if score > 90:
#     # This is excellent
# # 이렇게 하면 에러 발생

# Pass 사용 (정상 작동)
score = 85
if score > 90:
    pass  # This is excellent
print("Score processed")
````

````python
# 여러 조건에서 Pass 사용
value = 50
if value < 0:
    print("Negative value")
elif value == 0:
    pass  # Zero case - no action needed
else:
    print("Positive value")
````

## Java와의 비교

### 문법 차이

**Python:**
```python
# Python If Statement
a = 33
b = 200
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")

# Shorthand If ... Else (Ternary Operator)
result = "Good day" if time < 18 else "Good evening"

# Logical Operators
if a > b and c > a:
    print("Both conditions are True")
```

**Java:**
```java
// Java If Statement
int a = 33;
int b = 200;
if (b > a) {
    System.out.println("b is greater than a");
} else if (a == b) {
    System.out.println("a and b are equal");
} else {
    System.out.println("a is greater than b");
}

// Ternary Operator
String result = (time < 18) ? "Good day." : "Good evening.";

// Logical Operators
if (a > b && c > a) {
    System.out.println("Both conditions are true");
}
```

### 개념적 차이

- **들여쓰기 vs 중괄호**:
  - Python: 들여쓰기로 코드 블록 정의 (공백 또는 탭 사용)
  - Java: 중괄호 `{}`로 코드 블록 정의. 한 줄인 경우 중괄호 생략 가능하지만 권장하지 않음

- **조건식**:
  - Python: 조건 뒤에 콜론(`:`) 필요. 괄호는 선택사항
  - Java: 조건을 괄호 `()`로 감싸야 함. 중괄호로 블록 정의

- **Elif vs Else If**:
  - Python: `elif` 키워드 사용
  - Java: `else if` (두 단어) 사용

- **논리 연산자**:
  - Python: `and`, `or`, `not` (키워드)
  - Java: `&&`, `||`, `!` (기호)

- **Ternary Operator**:
  - Python: `value_if_true if condition else value_if_false` (조건이 중간)
  - Java: `condition ? value_if_true : value_if_false` (조건이 앞)

- **Truthiness**:
  - Python: 많은 값이 자동으로 True/False로 평가됨 (0, 빈 문자열, None 등은 False)
  - Java: 조건식은 반드시 boolean 타입이어야 함. Primitive 타입은 자동 변환 없음

- **Pass Statement**:
  - Python: `pass` 문으로 빈 블록 처리 (문법적으로 필요)
  - Java: 빈 블록은 그냥 `{}`로 처리. 별도의 키워드 없음

- **Boolean 변수 사용**:
  - Python: `if is_logged_in:` (직접 사용 가능)
  - Java: `if (isLoggedIn)` 또는 `if (isLoggedIn == true)` (괄호 필요)


## 정리

### 핵심 내용

1. **If Statement**: 조건이 True일 때 코드 블록 실행. Python은 들여쓰기로 블록 정의
2. **Elif Statement**: 여러 조건을 순차적으로 확인. 첫 번째 참인 조건만 실행
3. **Else Statement**: 모든 조건이 False일 때 실행되는 기본 동작 제공
4. **Shorthand If**: 한 줄로 간단한 if 문 작성 가능
5. **Ternary Operator**: `value_if_true if condition else value_if_false` 구문으로 조건부 값 할당
6. **Logical Operators**: `and`, `or`, `not`으로 조건 결합. 우선순위는 `not` > `and` > `or`
7. **Nested If**: if 문 안에 if 문을 중첩하여 복잡한 조건 처리. 논리 연산자로 단순화 가능한 경우도 있음
8. **Pass Statement**: 빈 블록을 위한 플레이스홀더. 개발 중 구조만 만들 때 유용
9. **Truthiness**: Python은 많은 값이 자동으로 True/False로 평가됨 (0, 빈 문자열, None 등은 False)
10. **Indentation**: Python의 핵심. 같은 블록 내에서는 일관된 들여쓰기 필수

### 느낀 점

- Python의 들여쓰기 기반 블록 정의가 깔끔하고 읽기 쉬움.
- Elif가 `else if`보다 간결하고 직관적.
- Ternary operator의 구문이 Java와 다르지만 익숙해지면 편리함.
- 논리 연산자가 키워드(`and`, `or`, `not`)라서 읽기 쉬움.
- Truthiness 개념이 있어서 조건문 작성이 유연함.
- Pass 문이 개발 중 구조만 잡을 때 매우 유용함.
